```python
def missing_values_per_column(df):
    """
    Esta funci√≥n analiza y muestra informaci√≥n sobre los valores faltantes en un DataFrame de pandas.
    - Imprime el n√∫mero de valores faltantes por columna.
    - Calcula y muestra el porcentaje de valores faltantes por columna.
    - Muestra el porcentaje total de valores faltantes en el DataFrame.
    - Genera un gr√°fico de barras que visualiza el porcentaje de valores faltantes por columna.

    Argumentos:
    df -- DataFrame de pandas que se analiza para encontrar valores faltantes.

    Retorno:
    None
    """

    # Paso 1: Encontrar columnas con valores faltantes y calcular la cantidad de valores faltantes por columna
    faltantes_por_columna = df.isnull().sum()
    
    # Paso 2: Filtrar las columnas que tienen valores faltantes y convertir los resultados a un diccionario
    data1 = faltantes_por_columna[faltantes_por_columna > 0]
    variables_con_faltantes = data1.to_dict()
    
    # Imprimir la cantidad de valores faltantes por columna en formato JSON con sangr√≠a para mejor lectura
    print('\nCant. valores faltantes por columna:',
          json.dumps(variables_con_faltantes, indent=4),
          sep="\n")
    
    # Paso 3: Calcular el porcentaje de valores faltantes por columna y formatearlo como porcentaje
    data = (data1 / len(df)) * 100
    data = data.to_dict()
    porcentaje_faltantes_por_columna = {key: f'{value:.2f}%' for key, value in data.items()}
    
    # Paso 4: Calcular el porcentaje total de valores faltantes en todo el DataFrame
    porcentaje_total_faltantes = (df.isnull().sum().sum() / df.size) * 100
    porcentaje_total_no_faltantes = (df.notnull().sum().sum() / df.size) * 100
    
    # Imprimir el porcentaje de valores faltantes por columna en formato JSON con sangr√≠a
    print('\n% valores faltantes por columna:', f'{json.dumps(porcentaje_faltantes_por_columna, indent=4)}', sep="\n")

    # Crear un gr√°fico de barras para visualizar el porcentaje de valores faltantes por columna
    plt.figure(figsize=(6, 3))
    plt.bar(variables_con_faltantes.keys(), variables_con_faltantes.values(), color=['blue', 'orange', 'green'])
    plt.title('Porcentaje de datos faltantes por columna')
    plt.ylabel('Porcentaje')
    plt.show()
    
    # Imprimir el porcentaje total de valores faltantes y no faltantes en el DataFrame
    print(f"\nPorcentaje total de valores faltantes en la base de datos: {porcentaje_total_faltantes:.2f}%")
    print(f"Porcentaje total de valores en la base de datos: {porcentaje_total_no_faltantes:.2f}%")

# Llamada a la funci√≥n con el DataFrame 'df'
missing_values_per_column(df)
```

#### **‚≠ïÔ∏è Imputaci√≥n por extracci√≥n de datos**

El proceso de imputaci√≥n de datos en este caso implica acceder a las URLs de las noticias y extraer informaci√≥n relevante para llenar los datos faltantes. Para aquellas noticias que no tienen t√≠tulo, se accede al enlace correspondiente y se extrae el t√≠tulo de la p√°gina. Adem√°s, se obtiene el primer p√°rrafo del contenido de la noticia desde el mismo enlace para completar las columnas de info y contenido en el conjunto de datos. Este enfoque permite enriquecer el conjunto de datos con informaci√≥n actualizada y espec√≠fica, mejorando su calidad y utilidad para an√°lisis posteriores o modelos de machine learning.

```python
# Configuraci√≥n del archivo de log
logging.basicConfig(
    filename='output_proceso_imputacion.log',  # Nombre del archivo donde se guardan los logs
    level=logging.INFO,  # Nivel de registro (INFO captura los mensajes de informaci√≥n est√°ndar)
    format='%(asctime)s - %(levelname)s - %(message)s',  # Formato de los mensajes en el archivo de log
)

def get_title_and_content(url, get_element):
    """
    Esta funci√≥n realiza una solicitud GET a una p√°gina web, analiza su contenido y devuelve el t√≠tulo o el contenido
    de la p√°gina seg√∫n el par√°metro 'get_element'. Los resultados y los errores se registran en un archivo de log.

    Argumentos:
    url -- URL de la p√°gina web a la que se realiza la solicitud.
    get_element -- Especifica qu√© parte de la p√°gina se desea obtener ('title' para el t√≠tulo o 'content' para el contenido).

    Retorno:
    str -- T√≠tulo o contenido de la p√°gina, o 'Sin Info' si hay un error o no se encuentra la informaci√≥n.
    """

    # Hacer una solicitud GET a la p√°gina web
    response = requests.get(url)

    # Registrar en el archivo de log la URL de la solicitud
    logging.info("-------------------------------------------------------------------------")
    logging.info(f'[URL Noticia]: {url}')
    logging.info("-------------------------------------------------------------------------")
    
    # Verificar que la solicitud fue exitosa (c√≥digo de estado 200)
    if response.status_code == 200:
        # Analizar el contenido de la p√°gina utilizando BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        response_data = ''  # Variable para almacenar el resultado de la solicitud
        
        # Si se solicita el t√≠tulo de la p√°gina
        if get_element == 'title':
            titulo = soup.title.string if soup.title else 'No se encontr√≥ el t√≠tulo'
            response_data = titulo
            # Registrar en el archivo de log el t√≠tulo encontrado
            logging.info(f'[T√≠tulo de la p√°gina]: {titulo}')
            logging.info("-------------------------------------------------------------------------")
            
        # Si se solicita el contenido de la p√°gina
        if get_element == 'content':
            # Buscar todos los elementos <div> con la clase 'paragraph'
            parrafos_divs = soup.find_all('div', class_='paragraph')
            # Unir el texto de todos los p√°rrafos encontrados
            contenido_parrafos = ' '.join([div.text.strip() for div in parrafos_divs if div.text.strip()]) if parrafos_divs else 'No se encontr√≥ el contenido del p√°rrafo'
            response_data = contenido_parrafos
            # Registrar en el archivo de log una muestra del contenido encontrado (hasta 50 caracteres)
            logging.info(f'[Contenido de la noticia]: {contenido_parrafos[:50]}')
            logging.info("-------------------------------------------------------------------------")
        
        return response_data
    else:
        # Registrar en el archivo de log el error si la solicitud no fue exitosa
        logging.error(f'Error al acceder a la p√°gina. C√≥digo de estado: {response.status_code}')
        return 'Sin Info'
```

#### **‚≠ïÔ∏è Imputaci√≥n de Titulos de noticias**

```python
# Variable de bandera para controlar si se deben imputar valores faltantes en la columna 'T√≠tulo'.
flag_imputer = False  # Para imputar valores, coloque esta bandera en True.

if flag_imputer:
    # Crear una copia del DataFrame original para no modificarlo directamente.
    df_new = df.copy()

    # Iterar sobre cada fila del DataFrame para verificar y completar valores faltantes en la columna 'T√≠tulo'.
    for index, row in df_new.iterrows():
        # Verificar si el valor de la columna 'T√≠tulo' es nulo o est√° vac√≠o.
        if pd.isnull(row['T√≠tulo']) or row['T√≠tulo'] == '':
            # Imputar el valor faltante llamando a la funci√≥n `get_title_and_content` para obtener el t√≠tulo desde la URL.
            df_new.at[index, 'T√≠tulo'] = get_title_and_content(df_new.at[index, 'Enlaces'], 'title')
    
    # Llamar a la funci√≥n `missing_values_per_column` para mostrar informaci√≥n sobre los valores faltantes en el DataFrame.
    missing_values_per_column(df_new)
```

#### **‚≠ïÔ∏è Imputaci√≥n de Informaci√≥n de noticias**

Solo se utilizar√° la cantidad promedio de caracteres de la columna info para imputar los datos faltantes, con el fin de evitar la inserci√≥n de una cantidad de informaci√≥n que pueda alterar los procesos de procesamiento de lenguaje natural (PLN). Al aplicar esta estrategia, se asegura que la informaci√≥n complementaria que se a√±ade no sea excesivamente extensa o inconsistente con el resto del contenido, manteniendo as√≠ la coherencia y calidad del conjunto de datos. Esto contribuye a que los modelos de PLN trabajen con datos que respeten una longitud promedio, evitando desviaciones que podr√≠an afectar el rendimiento y la precisi√≥n de los an√°lisis y predicciones.

```python
# Calcular el n√∫mero de caracteres promedio en la columna 'info'
promedio_caracteres = df['info'].str.len().mean()

# Imprimir el resultado con un mensaje formateado
print(f'El n√∫mero de caracteres promedio en la columna "info" es: {promedio_caracteres:.2f}')
```

```python
if flag_imputer:
    # Iterar sobre cada fila del DataFrame para verificar y completar valores faltantes en la columna 'info'.
    for index, row in df_new.iterrows():
        # Verificar si el valor de la columna 'info' es nulo o est√° vac√≠o.
        if pd.isnull(row['info']) or row['info'] == '':
            # Llamar a la funci√≥n `get_title_and_content` para obtener el contenido desde la URL en la columna 'Enlaces'.
            content = get_title_and_content(df_new.at[index, 'Enlaces'], 'content')
            
            # Truncar el contenido para que no exceda el promedio de caracteres permitido.
            if len(content) > promedio_caracteres:
                content = content[:int(promedio_caracteres)]  # Truncar el contenido al l√≠mite de caracteres especificado
            
            # Asignar el contenido truncado a la columna 'info' del DataFrame.
            df_new.at[index, 'info'] = content
    
    # Llamar a la funci√≥n `missing_values_per_column` para mostrar informaci√≥n sobre los valores faltantes en el DataFrame.
    missing_values_per_column(df_new)
```

#### **‚≠ïÔ∏è Imputaci√≥n de Informaci√≥n de Contenido**

```python
# Verificar si la bandera 'flag_imputer' est√° activada para imputar valores faltantes
if flag_imputer:
    # Iterar sobre cada fila del DataFrame 'df_new'
    for index, row in df_new.iterrows():
        # Comprobar si el valor en la columna 'contenido' es nulo o una cadena vac√≠a
        if pd.isnull(row['contenido']) or row['contenido'] == '':
            # Llamar a la funci√≥n 'get_title_and_content' para obtener el contenido de la p√°gina web
            content = get_title_and_content(df_new.at[index, 'Enlaces'], 'content')
            
            # Asignar el contenido recuperado a la columna 'contenido' en la fila correspondiente
            df_new.at[index, 'contenido'] = content
    
    # Llamar a la funci√≥n para verificar y mostrar los valores faltantes en el DataFrame despu√©s de la imputaci√≥n
    missing_values_per_column(df_new)
```

## ‚ùáÔ∏è Filtrado de datos

En la etapa de **Filtrado de Datos**, se busca optimizar la calidad del conjunto de datos mediante la exclusi√≥n de categor√≠as irrelevantes, como *‚Äúarchivo‚Äù*, que no aportan valor a los objetivos del an√°lisis. ## ‚ùáÔ∏è Preprocesamiento de Texto

Estrateg√≠as de procesamiento de texto a implementar:
- **Tokenizaci√≥n:** Separar el texto en palabras o tokens individuales.
- **Normalizaci√≥n:** Convertir el texto a min√∫sculas y eliminar caracteres no deseados.
- **Eliminaci√≥n de Stopwords:** Quitar palabras comunes que no aportan valor sem√°ntico.

```python
# Descargar recursos de nltk si no est√°n descargados
nltk.download('stopwords')
nltk.download('punkt')

# Definir las stopwords en ingl√©s (puedes cambiar el idioma si es necesario)
stop_words = set(stopwords.words('english'))

# Definir la funci√≥n de normalizaci√≥n y eliminaci√≥n de stopwords
def normalizar_texto(texto):
    # Convertir a min√∫sculas
    texto = texto.lower()
    # Eliminar caracteres no alfab√©ticos
    texto = re.sub(r'[^a-z\s]', '', texto)
    # Tokenizar el texto
    palabras = word_tokenize(texto)
    # Eliminar stopwords
    palabras = [palabra for palabra in palabras if palabra not in stop_words]
    # Reconstruir el texto sin las stopwords
    return ' '.join(palabras)

# Aplicar la funci√≥n de preprocesamiento al DataFrame
df_filtrado['Texto_Procesado'] = df_filtrado['contenido'].astype(str).apply(normalizar_texto)

# Tokenizaci√≥n y generaci√≥n de secuencias
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=10000)  # Define un vocabulario m√°ximo si es necesario
tokenizer.fit_on_texts(df_filtrado['Texto_Procesado'])
sequences = tokenizer.texts_to_sequences(df_filtrado['Texto_Procesado'])

# Calcular la longitud m√°xima de las secuencias
max_sequence_len = max([len(x) for x in sequences])

# Rellenar las secuencias para que tengan una longitud uniforme
sequences_padded = tf.keras.preprocessing.sequence.pad_sequences(sequences, maxlen=max_sequence_len)

# Mostrar una muestra de las secuencias procesadas
print(f"Secuencias procesadas (muestra): {sequences_padded[:5]}")
print(f"Secuencias procesadas (shape): {sequences_padded.shape}")
print(f"Dataframe (Etiquetas): {df_filtrado['Etiqueta'].unique().shape[0]}")
```

## ‚ùáÔ∏è Divisi√≥n del conjunto de datos

La tarea de divisi√≥n del conjunto de datos es fundamental para garantizar que el modelo se entrene y eval√∫e de manera efectiva. Para ello, se debe dividir el conjunto de datos preprocesado en dos partes: un subconjunto de entrenamiento y otro de validaci√≥n. Es importante realizar esta divisi√≥n de forma aleatoria para evitar sesgos, pero asegur√°ndose de que se mantenga una proporci√≥n similar de las categor√≠as en ambos conjuntos. De esta manera, se preserva la representaci√≥n de cada clase y se facilita una evaluaci√≥n m√°s precisa del rendimiento del modelo en datos no vistos, lo que contribuye a una mejor generalizaci√≥n.

```python
# Crear una codificaci√≥n num√©rica para las etiquetas
# Convertimos la columna 'Etiqueta' del DataFrame 'df_filtrado' en una representaci√≥n num√©rica.
# La funci√≥n 'astype('category').cat.codes' asigna un c√≥digo num√©rico a cada categor√≠a √∫nica.
df_filtrado['Etiqueta_Num'] = df_filtrado['Etiqueta'].astype('category').cat.codes

# Convertir las etiquetas a formato one-hot encoding
# Calculamos el n√∫mero total de clases √∫nicas en las categor√≠as seleccionadas.
num_classes = len(categorias_seleccionadas)
# Usamos 'tf.keras.utils.to_categorical' para convertir las etiquetas num√©ricas en una representaci√≥n one-hot.
# Esto crea una matriz de codificaci√≥n one-hot para las etiquetas de las noticias.
y = tf.keras.utils.to_categorical(df_filtrado['Etiqueta_Num'], num_classes=num_classes)

# Dividir los datos en conjuntos de entrenamiento y prueba manteniendo la proporci√≥n de categor√≠as
# Utilizamos 'train_test_split' para dividir las caracter√≠sticas (sequences_padded) y las etiquetas (y)
# en conjuntos de entrenamiento y prueba. La opci√≥n 'test_size=0.2' indica que el 20% de los datos se usan para prueba.
# 'random_state=42' asegura la reproducibilidad de la divisi√≥n, y 'stratify' mantiene la proporci√≥n de clases en ambas divisiones.
X_train, X_val, y_train, y_val = train_test_split(
    sequences_padded,  # Caracter√≠sticas de entrada
    y,  # Etiquetas one-hot codificadas
    test_size=0.2,  # Proporci√≥n de datos para el conjunto de prueba
    random_state=42,  # Semilla para reproducibilidad
    stratify=df_filtrado['Etiqueta_Num']  # Mantener proporciones de categor√≠as en la divisi√≥n
)

# Imprimir las formas de los conjuntos de datos
# Imprimimos las formas de los conjuntos de entrenamiento y prueba para verificar que la divisi√≥n se realiz√≥ correctamente.
print(f"X_train shape: {X_train.shape}")  # Forma de las caracter√≠sticas de entrenamiento
print(f"y_train shape: {y_train.shape}")  # Forma de las etiquetas de entrenamiento
print(f"X_test shape: {X_val.shape}")     # Forma de las caracter√≠sticas de prueba
print(f"y_test shape: {y_val.shape}")     # Forma de las etiquetas de prueba
```

#### **‚≠ïÔ∏è Importante**

- **Incorporaci√≥n de stratify:** Se a√±adi√≥ stratify=df_filtrado['Etiqueta_Num'] para que train_test_split divida los datos respetando la proporci√≥n de las categor√≠as en Etiqueta_Num.
- **Mantener aleatoriedad:** La aleatoriedad sigue siendo controlada por random_state=42.
- **Salida balanceada:** Los conjuntos de entrenamiento y prueba mantendr√°n distribuciones similares para cada categor√≠a.

## ‚ùáÔ∏è Implementaci√≥n de Modelo RNN

#### **‚≠ïÔ∏è Procesamiento Modelo RNN**

```python
# Preparar los datos
# Extraemos las caracter√≠sticas (texto procesado) y las etiquetas de las columnas relevantes del DataFrame.
X = df_filtrado['Texto_Procesado'].values  # Caracter√≠sticas de entrada: texto procesado
y = df_filtrado['Etiqueta'].values  # Etiquetas de salida

# Tokenizaci√≥n y secuenciaci√≥n de texto
# Se define el n√∫mero m√°ximo de palabras a considerar y la longitud m√°xima de las secuencias de texto.
max_words = 10000  # N√∫mero m√°ximo de palabras a considerar en el vocabulario
max_len = 150  # Longitud m√°xima de la secuencia de texto

# Se crea un tokenizador que convierte el texto en secuencias de enteros, limitando el vocabulario a 'max_words'.
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(X)  # Ajustamos el tokenizador a los textos de entrenamiento

# Convertimos los textos en secuencias de n√∫meros y las rellenamos para que tengan la longitud 'max_len'.
X_seq = tokenizer.texts_to_sequences(X)  # Convertir el texto en secuencias de enteros
X_pad = pad_sequences(X_seq, maxlen=max_len)  # Rellenar las secuencias a 'max_len'

# Convertir las etiquetas a categor√≠as num√©ricas y hacer una codificaci√≥n one-hot
# Usamos LabelEncoder para transformar las etiquetas de texto a valores num√©ricos.
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)  # Codificar las etiquetas num√©ricas
y_one_hot = to_categorical(y_encoded)  # Convertir a codificaci√≥n one-hot

# Construcci√≥n del modelo RNN
embedding_dim = 100  # Dimensi√≥n de los vectores de embedding para la capa de embedding

# Se define un modelo secuencial con las siguientes capas:
model_rnn = Sequential([
    Embedding(input_dim=max_words, output_dim=embedding_dim, input_length=max_len),  # Capa de embedding
    SimpleRNN(32, return_sequences=False),  # Capa RNN con 32 unidades, sin secuencias de salida
    Dense(len(label_encoder.classes_), activation='softmax')  # Capa de salida con activaci√≥n softmax para clasificaci√≥n multiclase
])

# Compilaci√≥n del modelo
# Se usa el optimizador 'adam' y la funci√≥n de p√©rdida 'categorical_crossentropy' para clasificaci√≥n multiclase.
model_rnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Divisi√≥n de datos (ejemplo, si no est√°n divididos)
# Dividimos los datos en conjuntos de entrenamiento y validaci√≥n, manteniendo la proporci√≥n de clases.
X_train, X_val, y_train, y_val = train_test_split(X_pad, y_one_hot, test_size=0.2, random_state=42, stratify=y_one_hot)

# Entrenamiento del modelo
# Entrenamos el modelo con los datos de entrenamiento, utilizando 20 √©pocas y un tama√±o de lote de 32.
history_rnn = model_rnn.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_val, y_val))

# Evaluaci√≥n del modelo
# Evaluamos el modelo en el conjunto de validaci√≥n para obtener la p√©rdida y precisi√≥n.
loss, accuracy = model_rnn.evaluate(X_val, y_val)

# Realizar predicciones en el conjunto de validaci√≥n
# Obtenemos las predicciones del modelo y convertimos las probabilidades en clases predichas.
y_pred = model_rnn.predict(X_val)
y_pred_classes = np.argmax(y_pred, axis=1)  # Obtener las clases predichas (√≠ndices de clase)
y_true = np.argmax(y_val, axis=1)  # Convertir las etiquetas one-hot a clases enteras

# Calcular recall y F1-score
# Calculamos el recall y el F1-score ponderados para evaluar la calidad de las predicciones.
recall = recall_score(y_true, y_pred_classes, average='weighted')
f1 = f1_score(y_true, y_pred_classes, average='weighted')

# Generar la matriz de confusi√≥n
# Calculamos la matriz de confusi√≥n para observar la cantidad de verdaderos positivos, falsos positivos, etc.
conf_matrix = confusion_matrix(y_true, y_pred_classes)

# Generar el reporte de clasificaci√≥n
# Generamos un reporte de clasificaci√≥n detallado que incluye precisi√≥n, recall y F1-score por clase.
class_report = classification_report(y_true, y_pred_classes, target_names=label_encoder.classes_)

# Imprimir resultados
# Imprimimos los resultados de la evaluaci√≥n, la matriz de confusi√≥n y el reporte de clasificaci√≥n.
print(f'P√©rdida: {loss:.4f}, Precisi√≥n: {accuracy:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}')
print("\nMatriz de Confusi√≥n:")
print(conf_matrix)
print("\nReporte de Clasificaci√≥n:")
print(class_report)
```

#### **‚≠ïÔ∏è Modelo RNN (Optimizado)**

- Regularizaci√≥n L2 en las capas RNN y densas para evitar el sobreajuste.
- Dropout del 30% en varias capas para mejorar la generalizaci√≥n.
- Capa densa oculta con activaci√≥n ReLU y regularizaci√≥n L2 para una mejor representaci√≥n de caracter√≠sticas.
- Uso de return_sequences=True en la primera capa RNN para mantener la secuencia y permitir la siguiente capa RNN a recibir datos secuenciales.

```python
# Se crea un tokenizador que convierte el texto en secuencias de enteros, limitando el vocabulario a 'max_words'.
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(X)  # Ajustamos el tokenizador a los textos de entrenamiento

# Convertimos los textos en secuencias de n√∫meros y las rellenamos para que tengan la longitud 'max_len'.
X_seq = tokenizer.texts_to_sequences(X)  # Convertir el texto en secuencias de enteros
X_pad = pad_sequences(X_seq, maxlen=max_len)  # Rellenar las secuencias a 'max_len'

# Convertir las etiquetas a categor√≠as num√©ricas y hacer una codificaci√≥n one-hot
# Usamos LabelEncoder para transformar las etiquetas de texto a valores num√©ricos.
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)  # Codificar las etiquetas num√©ricas
y_one_hot = to_categorical(y_encoded)  # Convertir a codificaci√≥n one-hot

# Construcci√≥n del modelo RNN mejorado con recomendaciones
embedding_dim = 100  # Dimensi√≥n de los vectores de embedding para la capa de embedding

# Se define un modelo secuencial con las siguientes capas:
model_rnn = Sequential([
    Embedding(input_dim=max_words, output_dim=embedding_dim, input_length=max_len),  # Capa de embedding con vectores de caracter√≠sticas
    SimpleRNN(64, return_sequences=True, kernel_regularizer='l2'),  # Primera capa RNN con regularizaci√≥n L2 y retorno de secuencias activado
    Dropout(0.3),  # Capa de abandono (dropout) para prevenir sobreajuste con tasa del 30%
    SimpleRNN(64, return_sequences=False, kernel_regularizer='l2'),  # Segunda capa RNN con regularizaci√≥n L2 y retorno de secuencias desactivado
    Dropout(0.3),  # Otra capa de abandono (dropout) para prevenir sobreajuste
    Dense(128, activation='relu', kernel_regularizer='l2'),  # Capa densa oculta con activaci√≥n ReLU y regularizaci√≥n L2
    Dropout(0.3),  # Capa de abandono (dropout) para evitar sobreajuste
    Dense(len(label_encoder.classes_), activation='softmax')  # Capa de salida con activaci√≥n softmax para clasificaci√≥n multiclase
])

# Compilaci√≥n del modelo
# Se usa el optimizador 'adam' y la funci√≥n de p√©rdida 'categorical_crossentropy' para clasificaci√≥n multiclase.
model_rnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Divisi√≥n de datos (ejemplo, si no est√°n divididos)
# Dividimos los datos en conjuntos de entrenamiento y validaci√≥n, manteniendo la proporci√≥n de clases.
X_train, X_val, y_train, y_val = train_test_split(X_pad, y_one_hot, test_size=0.2, random_state=42, stratify=y_one_hot)

# Entrenamiento del modelo
# Entrenamos el modelo con los datos de entrenamiento, utilizando 20 √©pocas y un tama√±o de lote de 32.
history_rnn = model_rnn.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_val, y_val))

# Evaluaci√≥n del modelo
# Evaluamos el modelo en el conjunto de validaci√≥n para obtener la p√©rdida y precisi√≥n.
loss, accuracy = model_rnn.evaluate(X_val, y_val)

# Realizar predicciones en el conjunto de validaci√≥n
# Obtenemos las predicciones del modelo y convertimos las probabilidades en clases predichas.
y_pred = model_rnn.predict(X_val)
y_pred_classes = np.argmax(y_pred, axis=1)  # Obtener las clases predichas (√≠ndices de clase)
y_true = np.argmax(y_val, axis=1)  # Convertir las etiquetas one-hot a clases enteras

# Calcular recall y F1-score
# Calculamos el recall y el F1-score ponderados para evaluar la calidad de las predicciones.
recall = recall_score(y_true, y_pred_classes, average='weighted')
f1 = f1_score(y_true, y_pred_classes, average='weighted')

# Generar la matriz de confusi√≥n
# Calculamos la matriz de confusi√≥n para observar la cantidad de verdaderos positivos, falsos positivos, etc.
conf_matrix = confusion_matrix(y_true, y_pred_classes)

# Generar el reporte de clasificaci√≥n
# Generamos un reporte de clasificaci√≥n detallado que incluye precisi√≥n, recall y F1-score por clase.
class_report = classification_report(y_true, y_pred_classes, target_names=label_encoder.classes_)

# Imprimir resultados
# Imprimimos los resultados de la evaluaci√≥n, la matriz de confusi√≥n y el reporte de clasificaci√≥n.
print(f'P√©rdida: {loss:.4f}, Precisi√≥n: {accuracy:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}')
print("\nMatriz de Confusi√≥n:")
print(conf_matrix)
print("\nReporte de Clasificaci√≥n:")
print(class_report)
```

#### **‚≠ïÔ∏è Mejora de Rendimiento de Modelo LSTM**

Para mejorar el rendimiento del modelo LSTM se debe evitar el sobreajuste y optimizar el entrenamiento del modelo LSTM.

#### Cambios realizados en el modelo LSTM

- **LSTM Bidireccional**: Se a√±adi√≥ para permitir que el modelo capture dependencias de la secuencia en ambas direcciones (izquierda a derecha y derecha a izquierda), mejorando la comprensi√≥n de la informaci√≥n temporal.

- **Regularizaci√≥n L2**: Se incorpor√≥ en las capas LSTM y densas para reducir el sobreajuste penalizando los pesos grandes, ayudando a mejorar la generalizaci√≥n del modelo.

- **ReduceLROnPlateau**: Se agreg√≥ como un callback para ajustar la tasa de aprendizaje autom√°ticamente si la p√©rdida de validaci√≥n no mejora en un n√∫mero especificado de √©pocas, lo que puede ayudar a evitar que el modelo quede atrapado en m√≠nimos locales.

- **M√°s capas y unidades**: Se increment√≥ la complejidad del modelo a√±adiendo una segunda capa LSTM y unidades de mayor capacidad para mejorar la capacidad de aprendizaje del modelo.

- **Tasa de aprendizaje personalizada**: Se estableci√≥ un valor de 0.001 en el optimizador Adam, lo cual es m√°s com√∫n para modelos LSTM, permitiendo una mejor convergencia.

- **Callback de EarlyStopping**: Se mantuvo con una paciencia de 10 √©pocas para detener el entrenamiento si no hay mejora y restaurar los mejores pesos encontrados.

```python
# Configuraci√≥n de par√°metros para la tokenizaci√≥n y el modelo
max_words = 10000  # N√∫mero m√°ximo de palabras a considerar en el vocabulario
max_len = 150      # Longitud m√°xima de las secuencias de entrada (m√°ximo de palabras por muestra)

# Definici√≥n del modelo LSTM mejorado con regularizaci√≥n y capas bidireccionales
model_lstm = Sequential([
    # Capa de embedding para convertir palabras en vectores de caracter√≠sticas
    Embedding(input_dim=max_words, output_dim=100, input_length=max_len),
    
    # Capa LSTM bidireccional con 64 unidades, retorno de secuencias activado y regularizaci√≥n recurrente
    Bidirectional(LSTM(64, return_sequences=True, recurrent_dropout=0.3)),
    Dropout(0.5),  # Capa de abandono para prevenir el sobreajuste
    
    # Capa LSTM adicional con 32 unidades y sin retorno de secuencias, con regularizaci√≥n recurrente
    LSTM(32, return_sequences=False, recurrent_dropout=0.3),
    Dropout(0.5),  # Capa de abandono para mayor regularizaci√≥n
    
    # Capa densa con 128 unidades y activaci√≥n ReLU
    Dense(128, activation='relu'),
    Dropout(0.5),  # Regularizaci√≥n adicional
    
    # Capa de salida con activaci√≥n softmax para clasificaci√≥n multiclase
    Dense(len(label_encoder.classes_), activation='softmax')
])

# Compilaci√≥n del modelo con el optimizador 'adam', funci√≥n de p√©rdida 'categorical_crossentropy' y m√©trica de 'accuracy'
model_lstm.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Configuraci√≥n de callbacks para controlar el entrenamiento
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True)

# Entrenamiento del modelo con callbacks para mejorar la convergencia y guardar el mejor modelo
history_lstm_optimized = model_lstm.fit(
    X_train, y_train,
    epochs=20,  # Aumentar el n√∫mero de √©pocas para un entrenamiento m√°s exhaustivo
    batch_size=32,  # Tama√±o del lote de entrenamiento
    validation_data=(X_val, y_val),  # Datos de validaci√≥n para monitorear el desempe√±o
    callbacks=[early_stopping, model_checkpoint]  # Usar callbacks para control de entrenamiento
)

# Evaluaci√≥n del modelo en el conjunto de validaci√≥n
val_loss, val_accuracy = model_lstm.evaluate(X_val, y_val)

# Realizar predicciones en el conjunto de validaci√≥n
y_pred = model_lstm.predict(X_val)
y_pred_classes = np.argmax(y_pred, axis=1)  # Convertir las predicciones one-hot a clases enteras
y_true = np.argmax(y_val, axis=1)  # Convertir las etiquetas one-hot de validaci√≥n a clases enteras

# Calcular recall y F1-score para evaluar el rendimiento del modelo
recall = recall_score(y_true, y_pred_classes, average='weighted')
f1 = f1_score(y_true, y_pred_classes, average='weighted')

# Generar la matriz de confusi√≥n para evaluar la precisi√≥n de las predicciones por clase
conf_matrix = confusion_matrix(y_true, y_pred_classes)

# Generar un reporte de clasificaci√≥n con detalles por clase
class_report = classification_report(y_true, y_pred_classes, target_names=label_encoder.classes_)

# Imprimir los resultados de la evaluaci√≥n
print(f'P√©rdida: {val_loss:.4f}, Precisi√≥n: {val_accuracy:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}')
print("\nMatriz de Confusi√≥n:")
print(conf_matrix)
print("\nReporte de Clasificaci√≥n:")
print(class_report)
```

```python
# Configuraci√≥n del dispositivo
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Dataset personalizado
class NewsDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}, torch.tensor(int(self.labels[idx]))

# Filtrar el DataFrame para eliminar las filas con las etiquetas "pol√≠tica" y "salud"
df_filtrado = df_filtrado[~df_filtrado['Etiqueta'].isin(['politica', 'salud'])]

# Actualizar las etiquetas num√©ricas si es necesario
df_filtrado['Etiqueta_Num'] = df_filtrado['Etiqueta'].astype('category').cat.codes

# Preparaci√≥n de datos: extraer y procesar los textos y las etiquetas
train_texts = df_filtrado['Texto_Procesado'].tolist()
train_texts = [text[:512] for text in train_texts]  # Limitar cada texto a un m√°ximo de 512 caracteres
train_labels = df_filtrado['Etiqueta_Num'].tolist()

# Validaci√≥n de las etiquetas para asegurar que est√©n en el rango esperado
unique_labels = set(train_labels)
num_classes = len(unique_labels)
if not all(label < num_classes for label in unique_labels):
    raise ValueError(f"Encontradas etiquetas fuera del rango 0-{num_classes-1}")

# Inicializaci√≥n del tokenizador BERT para tokenizar los textos
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# Tokenizaci√≥n de los textos con truncamiento y padding, y un m√°ximo de 512 caracteres por texto
train_encodings = tokenizer(train_texts, truncation=True, padding=True, max_length=512)

# Preparaci√≥n de los datasets y dataloaders para entrenamiento y validaci√≥n
train_dataset = NewsDataset(train_encodings, train_labels)
train_size = int(0.8 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(train_dataset, [train_size, val_size])

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False)

# Carga y configuraci√≥n del modelo BERT para clasificaci√≥n de secuencias
model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=num_classes).to(device)

# Definici√≥n del optimizador con una tasa de aprendizaje ajustada
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)

# Entrenamiento del modelo
model.train()
train_loss_history = []  # Lista para almacenar la p√©rdida de cada √©poca
val_loss_history = []  # Lista para almacenar la p√©rdida de validaci√≥n por √©poca
train_accuracy_history = []  # Lista para almacenar la precisi√≥n de entrenamiento por √©poca
val_accuracy_history = []  # Lista para almacenar la precisi√≥n de validaci√≥n por √©poca

for epoch in range(3):  # N√∫mero de √©pocas de entrenamiento
    total_loss = 0
    correct_predictions = 0
    total_samples = 0

    # Entrenamiento
    for batch in train_loader:
        inputs, labels = batch
        inputs = {key: val.to(device) for key, val in inputs.items()}
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(**inputs, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        # C√°lculo de la precisi√≥n de entrenamiento
        preds = torch.argmax(outputs.logits, axis=-1)
        correct_predictions += (preds == labels).sum().item()
        total_samples += labels.size(0)

    # Almacenar la p√©rdida total y la precisi√≥n de la √©poca de entrenamiento
    train_loss_history.append(total_loss)
    train_accuracy_history.append(correct_predictions / total_samples)

    # Evaluaci√≥n de la p√©rdida de validaci√≥n y la precisi√≥n de validaci√≥n
    model.eval()
    val_loss = 0
    correct_val_predictions = 0
    val_samples = 0
    with torch.no_grad():
        for batch in val_loader:
            inputs, labels = batch
            inputs = {key: val.to(device) for key, val in inputs.items()}
            labels = labels.to(device)

            outputs = model(**inputs, labels=labels)
            val_loss += outputs.loss.item()

            # C√°lculo de la precisi√≥n de validaci√≥n
            preds = torch.argmax(outputs.logits, axis=-1)
            correct_val_predictions += (preds == labels).sum().item()
            val_samples += labels.size(0)

    val_loss_history.append(val_loss)
    val_accuracy = correct_val_predictions / val_samples
    val_accuracy_history.append(val_accuracy)
    model.train()

    print(f"√âpoca {epoch+1}, P√©rdida de entrenamiento: {total_loss:.4f}, Precisi√≥n de entrenamiento: {train_accuracy_history[-1]:.4f}, P√©rdida de validaci√≥n: {val_loss:.4f}, Precisi√≥n de validaci√≥n: {val_accuracy:.4f}")

# Evaluaci√≥n del modelo en el conjunto de entrenamiento (o validaci√≥n)
model.eval()
predictions, true_labels = [], []
with torch.no_grad():
    for batch in train_loader:
        inputs, labels = batch
        inputs = {key: val.to(device) for key, val in inputs.items()}
        labels = labels.to(device)
        outputs = model(**inputs)
        preds = torch.argmax(outputs.logits, axis=-1).cpu().numpy()
        predictions.extend(preds)
        true_labels.extend(labels.cpu().numpy())

# C√°lculo de las m√©tricas de rendimiento
accuracy = accuracy_score(true_labels, predictions)
f1 = f1_score(true_labels, predictions, average="weighted")
recall = recall_score(true_labels, predictions, average='weighted')
validation_loss = sum(val_loss_history) / len(val_loss_history)

# Impresi√≥n de resultados finales
print(f"P√©rdida de validaci√≥n: {validation_loss:.4f}, Precisi√≥n: {accuracy:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}")
print("\nMatriz de Confusi√≥n:")
print(confusion_matrix(true_labels, predictions))
print("\nReporte de Clasificaci√≥n:")
print(classification_report(true_labels, predictions))
```

### ‚≠ïÔ∏è **Implementaci√≥n Modelo BERT** `mrm8488/bert-base-spanish-wwm-uncased` (Optimizado)

Para la optimizaci√≥n del modelo se practican:

1. **Regularizaci√≥n con Dropout**  
   Se agrega dropout en las capas de BERT para prevenir el sobreajuste.

2. **Weight Decay**  
   Se aplica regularizaci√≥n L2 en el optimizador `AdamW` para reducir el sobreajuste.

3. **Programaci√≥n de Tasa de Aprendizaje**  
   Se utiliza un programador (`lr_scheduler`) para ajustar din√°micamente la tasa de aprendizaje durante el entrenamiento.

4. **Aumento del Tama√±o del Lote**  
   Un lote m√°s grande puede estabilizar las actualizaciones del gradiente.

5. **Aumento de √âpocas**  
   M√°s √©pocas pueden ayudar a que el modelo converja mejor, aunque se debe monitorizar el sobreajuste.

6. **Gradiente Clipping**  
   Se limita el gradiente para evitar explosiones durante el entrenamiento.

7. **Validaci√≥n Temprana**  
   Se agrega validaci√≥n temprana para detener el entrenamiento si no hay mejora.

```python
# Configuraci√≥n de hiperpar√°metros
num_epochs = 5  # Incrementar el n√∫mero de √©pocas
batch_size = 16  # Aumentar el tama√±o del lote
learning_rate = 2e-5  # Reducir la tasa de aprendizaje
weight_decay = 0.01  # Regularizaci√≥n L2

# Carga y configuraci√≥n del modelo BERT para clasificaci√≥n de secuencias
model_optimized = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=num_classes).to(device)

# Modificaci√≥n del DataLoader con un tama√±o de lote m√°s grande
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

# Definir el optimizador con regularizaci√≥n L2 (Weight Decay)
optimizer = torch.optim.AdamW(model_optimized.parameters(), lr=learning_rate, weight_decay=weight_decay)

# Programaci√≥n de tasa de aprendizaje
num_training_steps = len(train_loader) * num_epochs
lr_scheduler = get_scheduler("linear", optimizer=optimizer, num_warmup_steps=0, num_training_steps=num_training_steps)

# Gradiente Clipping
max_grad_norm = 1.0

# Inicializar listas para almacenar los resultados por √©poca
train_loss_history = []
train_accuracy_history = []
val_loss_history = []
val_accuracy_history = []

# Inicializar listas para almacenar las predicciones y etiquetas para m√©tricas finales
true_labels = []
predictions = []

# Entrenamiento con regularizaci√≥n
best_val_loss = float("inf")
patience = 4  # N√∫mero de √©pocas sin mejora para detener
early_stop_counter = 0

model_optimized.train()
for epoch in range(num_epochs):
    total_loss = 0
    correct_predictions = 0
    total_samples = 0

    # Entrenamiento
    for batch in train_loader:
        inputs, labels = batch
        inputs = {key: val.to(device) for key, val in inputs.items()}
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model_optimized(**inputs, labels=labels)
        loss = outputs.loss
        loss.backward()

        # Aplicar Gradiente Clipping
        torch.nn.utils.clip_grad_norm_(model_optimized.parameters(), max_grad_norm)

        optimizer.step()
        lr_scheduler.step()

        total_loss += loss.item()

        # C√°lculo de la precisi√≥n de entrenamiento
        preds = torch.argmax(outputs.logits, axis=-1)
        correct_predictions += (preds == labels).sum().item()
        total_samples += labels.size(0)

        # Almacenar las etiquetas y predicciones para m√©tricas finales
        true_labels.extend(labels.cpu().numpy())
        predictions.extend(preds.cpu().numpy())

    # Calcular p√©rdida y precisi√≥n de entrenamiento
    train_loss = total_loss / len(train_loader)
    train_accuracy = correct_predictions / total_samples

    # Almacenar resultados de entrenamiento
    train_loss_history.append(train_loss)
    train_accuracy_history.append(train_accuracy)

    # Evaluaci√≥n en validaci√≥n
    model_optimized.eval()
    val_loss = 0
    correct_val_predictions = 0
    val_samples = 0
    with torch.no_grad():
        for batch in val_loader:
            inputs, labels = batch
            inputs = {key: val.to(device) for key, val in inputs.items()}
            labels = labels.to(device)

            outputs = model_optimized(**inputs, labels=labels)
            val_loss += outputs.loss.item()

            preds = torch.argmax(outputs.logits, axis=-1)
            correct_val_predictions += (preds == labels).sum().item()
            val_samples += labels.size(0)

    val_loss /= len(val_loader)
    val_accuracy = correct_val_predictions / val_samples

    # Almacenar resultados de validaci√≥n
    val_loss_history.append(val_loss)
    val_accuracy_history.append(val_accuracy)

    # Imprimir resultados de la √©poca
    print(f"√âpoca {epoch+1}, P√©rdida de entrenamiento: {train_loss:.4f}, Precisi√≥n de entrenamiento: {train_accuracy:.4f}, P√©rdida de validaci√≥n: {val_loss:.4f}, Precisi√≥n de validaci√≥n: {val_accuracy:.4f}")

    # Early Stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        early_stop_counter = 0
        torch.save(model_optimized.state_dict(), "best_model.pt")
    else:
        early_stop_counter += 1
        if early_stop_counter >= patience:
            print("Entrenamiento detenido por early stopping.")
            break

    model_optimized.train()

# Cargar el mejor modelo
model_optimized.load_state_dict(torch.load("best_model.pt"))

# C√°lculo de las m√©tricas de rendimiento
accuracy = accuracy_score(true_labels, predictions)
f1 = f1_score(true_labels, predictions, average="weighted")
recall = recall_score(true_labels, predictions, average='weighted')
validation_loss = sum(val_loss_history) / len(val_loss_history)

# Impresi√≥n de resultados finales
print(f"P√©rdida de validaci√≥n: {validation_loss:.4f}, Precisi√≥n: {accuracy:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}")
print("\nMatriz de Confusi√≥n:")
print(confusion_matrix(true_labels, predictions))
print("\nReporte de Clasificaci√≥n:")
print(classification_report(true_labels, predictions))
```
