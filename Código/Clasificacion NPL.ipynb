```python
def missing_values_per_column(df):
    # Escribir las variables que tienen datos faltantes, adem√°s, proporcionar el porcentaje total de valores faltantes en la base de datos.
    # Paso 1: Encontrar columnas con valores faltantes
    faltantes_por_columna = df.isnull().sum()
    
    # Paso 2: Filtrar columnas que tienen valores faltantes
    data1 = faltantes_por_columna[faltantes_por_columna > 0]
    variables_con_faltantes = data1.to_dict()
    
    print('\nCant. valores faltantes por columna:',
          json.dumps(variables_con_faltantes, indent=4),
          sep="\n")
    
    # Paso 3: Calcular el porcentaje de valores faltantes por columna
    data = (data1 / len(df)) * 100
    data = data.to_dict()
    porcentaje_faltantes_por_columna = {key: f'{value:.2f}%' for key, value in data.items()}
    
    # Paso 4: Calcular el porcentaje total de valores faltantes en el DataFrame
    porcentaje_total_faltantes = (df.isnull().sum().sum() / df.size) * 100
    porcentaje_total_no_faltantes = (df.notnull().sum().sum() / df.size) * 100
    
    # Resultados
    print('\n% valores faltantes por columna:', f'{json.dumps(porcentaje_faltantes_por_columna, indent=4)}', sep="\n")

    # Crear gr√°fico de barras
    plt.figure(figsize=(6, 3))
    plt.bar(variables_con_faltantes.keys(), variables_con_faltantes.values(), color=['blue', 'orange', 'green'])
    plt.title('Porcentaje de datos faltantes por columna')
    plt.ylabel('Porcentaje')
    plt.show()
    
    print(f"\nPorcentaje total de valores faltantes en la base de datos: {porcentaje_total_faltantes:.2f}%")
    print(f"Porcentaje total de valores en la base de datos: {porcentaje_total_no_faltantes:.2f}%")

missing_values_per_column(df)
```

#### **‚≠ïÔ∏è Imputaci√≥n por extracci√≥n de datos**

El proceso de imputaci√≥n de datos en este caso implica acceder a las URLs de las noticias y extraer informaci√≥n relevante para llenar los datos faltantes. Para aquellas noticias que no tienen t√≠tulo, se accede al enlace correspondiente y se extrae el t√≠tulo de la p√°gina. Adem√°s, se obtiene el primer p√°rrafo del contenido de la noticia desde el mismo enlace para completar las columnas de info y contenido en el conjunto de datos. Este enfoque permite enriquecer el conjunto de datos con informaci√≥n actualizada y espec√≠fica, mejorando su calidad y utilidad para an√°lisis posteriores o modelos de machine learning.

```python
# Configurar el archivo de log
logging.basicConfig(
    filename='output_proceso_imputacion.log',  # Nombre del archivo de log
    level=logging.INFO,  # Nivel de registro (INFO captura las impresiones est√°ndar)
    format='%(asctime)s - %(levelname)s - %(message)s',  # Formato del mensaje
)

def get_title_and_content(url, get_element):
    # Hacer una solicitud GET a la p√°gina web
    response = requests.get(url)

    logging.info("-------------------------------------------------------------------------")
    logging.info(f'[URL Noticia]: {url}')
    logging.info("-------------------------------------------------------------------------")
    
    # Verificar que la solicitud fue exitosa
    if response.status_code == 200:
        # Analizar el contenido de la p√°gina con BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        response_data = ''
        
        if get_element == 'title':
            # Buscar el t√≠tulo de la p√°gina
            titulo = soup.title.string if soup.title else 'No se encontr√≥ el t√≠tulo'
            response_data = titulo
            logging.info(f'[T√≠tulo de la p√°gina]: {titulo}')
            logging.info("-------------------------------------------------------------------------")
            
        if get_element == 'content':
            # Buscar todos los <div> con la clase 'paragraph'
            parrafos_divs = soup.find_all('div', class_='paragraph')
            contenido_parrafos = ' '.join([div.text.strip() for div in parrafos_divs if div.text.strip()]) if parrafos_divs else 'No se encontr√≥ el contenido del p√°rrafo'
            response_data = contenido_parrafos
            logging.info(f'[Contenido de la noticia]: {contenido_parrafos[:50]}')
            logging.info("-------------------------------------------------------------------------")
        
        return response_data
    else:
        logging.error(f'Error al acceder a la p√°gina. C√≥digo de estado: {response.status_code}')
        return 'Sin Info'
```

## ‚ùáÔ∏è Preprocesamiento de Texto

Estrateg√≠as de procesamiento de texto a implementar:
- **Tokenizaci√≥n:** Separar el texto en palabras o tokens individuales.
- **Normalizaci√≥n:** Convertir el texto a min√∫sculas y eliminar caracteres no deseados.
- **Eliminaci√≥n de Stopwords:** Quitar palabras comunes que no aportan valor sem√°ntico.

```python
# Descargar recursos de nltk si no est√°n descargados
nltk.download('stopwords')
nltk.download('punkt')

# Definir las stopwords en ingl√©s (puedes cambiar el idioma si es necesario)
stop_words = set(stopwords.words('english'))

# Definir la funci√≥n de normalizaci√≥n y eliminaci√≥n de stopwords
def normalizar_texto(texto):
    # Convertir a min√∫sculas
    texto = texto.lower()
    # Eliminar caracteres no alfab√©ticos
    texto = re.sub(r'[^a-z\s]', '', texto)
    # Tokenizar el texto
    palabras = word_tokenize(texto)
    # Eliminar stopwords
    palabras = [palabra for palabra in palabras if palabra not in stop_words]
    # Reconstruir el texto sin las stopwords
    return ' '.join(palabras)

# Aplicar la funci√≥n de preprocesamiento al DataFrame
df_filtrado['Texto_Procesado'] = df_filtrado['contenido'].astype(str).apply(normalizar_texto)

# Tokenizaci√≥n y generaci√≥n de secuencias
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=10000)  # Define un vocabulario m√°ximo si es necesario
tokenizer.fit_on_texts(df_filtrado['Texto_Procesado'])
sequences = tokenizer.texts_to_sequences(df_filtrado['Texto_Procesado'])

# Calcular la longitud m√°xima de las secuencias
max_sequence_len = max([len(x) for x in sequences])

# Rellenar las secuencias para que tengan una longitud uniforme
sequences_padded = tf.keras.preprocessing.sequence.pad_sequences(sequences, maxlen=max_sequence_len)

# Mostrar una muestra de las secuencias procesadas
print(f"Secuencias procesadas (muestra): {sequences_padded[:5]}")
print(f"Secuencias procesadas (shape): {sequences_padded.shape}")
print(f"Dataframe (Etiquetas): {df_filtrado['Etiqueta'].unique().shape[0]}")
```

## ‚ùáÔ∏è Divisi√≥n del conjunto de datos

La tarea de divisi√≥n del conjunto de datos es fundamental para garantizar que el modelo se entrene y eval√∫e de manera efectiva. Para ello, se debe dividir el conjunto de datos preprocesado en dos partes: un subconjunto de entrenamiento y otro de validaci√≥n. Es importante realizar esta divisi√≥n de forma aleatoria para evitar sesgos, pero asegur√°ndose de que se mantenga una proporci√≥n similar de las categor√≠as en ambos conjuntos. De esta manera, se preserva la representaci√≥n de cada clase y se facilita una evaluaci√≥n m√°s precisa del rendimiento del modelo en datos no vistos, lo que contribuye a una mejor generalizaci√≥n.

```python
# Preparar los datos
X = df_filtrado['Texto_Procesado'].values
y = df_filtrado['Etiqueta'].values

# Tokenizaci√≥n y secuenciaci√≥n de texto
max_words = 5000  # N√∫mero m√°ximo de palabras a considerar
max_len = 150  # Longitud m√°xima de la secuencia de texto

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(X)
X_seq = tokenizer.texts_to_sequences(X)
X_pad = pad_sequences(X_seq, maxlen=max_len)

# Convertir las etiquetas a categor√≠as num√©ricas y hacer una codificaci√≥n one-hot
#from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_one_hot = to_categorical(y_encoded)

# Construcci√≥n del modelo RNN
embedding_dim = 100  # Dimensi√≥n de los vectores de embedding

model_rnn = Sequential([
    Embedding(input_dim=max_words, output_dim=embedding_dim, input_length=max_len),
    SimpleRNN(32, return_sequences=False),
    Dense(len(label_encoder.classes_), activation='softmax')  # Capa de salida con activaci√≥n softmax para clasificaci√≥n multiclase
])

# Compilaci√≥n del modelo
model_rnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Divisi√≥n de datos (ejemplo, si no est√°n divididos)
X_train, X_val, y_train, y_val = train_test_split(X_pad, y_one_hot, test_size=0.2, random_state=42, stratify=y_one_hot)

# Entrenamiento del modelo
history_rnn = model_rnn.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_val, y_val))

# Evaluaci√≥n del modelo
loss, accuracy = model_rnn.evaluate(X_val, y_val)

# Realizar predicciones en el conjunto de validaci√≥n
y_pred = model_rnn.predict(X_val)
y_pred_classes = np.argmax(y_pred, axis=1)  # Obtener las clases predichas
y_true = np.argmax(y_val, axis=1)  # Convertir las etiquetas one-hot a clases enteras

# Calcular recall y F1-score
recall = recall_score(y_true, y_pred_classes, average='weighted')
f1 = f1_score(y_true, y_pred_classes, average='weighted')

# Generar la matriz de confusi√≥n
conf_matrix = confusion_matrix(y_true, y_pred_classes)

# Generar el reporte de clasificaci√≥n
class_report = classification_report(y_true, y_pred_classes, target_names=label_encoder.classes_)

# Imprimir resultados
print(f'P√©rdida: {loss:.4f}, Precisi√≥n: {accuracy:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}')
print("\nMatriz de Confusi√≥n:")
print(conf_matrix)
print("\nReporte de Clasificaci√≥n:")
print(class_report)
```

- **Tasa de aprendizaje personalizada**: Se estableci√≥ un valor de 0.001 en el optimizador Adam, lo cual es m√°s com√∫n para modelos LSTM, permitiendo una mejor convergencia.

- **Callback de EarlyStopping**: Se mantuvo con una paciencia de 10 √©pocas para detener el entrenamiento si no hay mejora y restaurar los mejores pesos encontrados.

```python
# Configuraci√≥n de par√°metros
max_words = 10000  # N√∫mero m√°ximo de palabras en el vocabulario
max_len = 150      # Longitud m√°xima de las secuencias de entrada

# Definici√≥n del modelo LSTM mejorado
model_lstm = Sequential([
    Embedding(input_dim=max_words, output_dim=100, input_length=max_len),
    Bidirectional(LSTM(64, return_sequences=True, recurrent_dropout=0.3)),  # LSTM bidireccional con regularizaci√≥n recurrente
    Dropout(0.5),  # Regularizaci√≥n para prevenir el sobreajuste
    LSTM(32, return_sequences=False, recurrent_dropout=0.3),  # Capa LSTM adicional
    Dropout(0.5),  # Otra capa de regularizaci√≥n
    Dense(128, activation='relu'),
    Dropout(0.5),  # Regularizaci√≥n adicional
    Dense(len(label_encoder.classes_), activation='softmax')  # Capa de salida para clasificaci√≥n multiclase
])

# Compilaci√≥n del modelo con un optimizador ajustado
model_lstm.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Configuraci√≥n de callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True)

# Entrenamiento del modelo
history_lstm_optimized = model_lstm.fit(
    X_train, y_train,
    epochs=20,  # Aumentar el n√∫mero de √©pocas para un mejor entrenamiento
    batch_size=32,
    validation_data=(X_val, y_val),
    callbacks=[early_stopping, model_checkpoint]  # Usar callbacks
)

# Evaluaci√≥n del modelo con el conjunto de validaci√≥n
val_loss, val_accuracy = model_lstm.evaluate(X_val, y_val)

# Realizar predicciones en el conjunto de validaci√≥n
y_pred = model_lstm.predict(X_val)
y_pred_classes = np.argmax(y_pred, axis=1)  # Obtener las clases predichas
y_true = np.argmax(y_val, axis=1)  # Convertir las etiquetas one-hot a clases enteras

# Calcular recall y F1-score
recall = recall_score(y_true, y_pred_classes, average='weighted')
f1 = f1_score(y_true, y_pred_classes, average='weighted')

# Generar la matriz de confusi√≥n
conf_matrix = confusion_matrix(y_true, y_pred_classes)

# Generar el reporte de clasificaci√≥n
class_report = classification_report(y_true, y_pred_classes, target_names=label_encoder.classes_)

# Imprimir resultados
print(f'P√©rdida: {val_loss:.4f}, Precisi√≥n: {val_accuracy:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}')
print("\nMatriz de Confusi√≥n:")
print(conf_matrix)
print("\nReporte de Clasificaci√≥n:")
print(class_report)
```

## ‚ùáÔ∏è Implementaci√≥n y entrenamiento de BERT

Para abordar tareas de clasificaci√≥n de texto en espa√±ol, se utilizan modelos de lenguaje avanzados como BERT, que permiten capturar el contexto de las palabras y mejorar la comprensi√≥n del texto. En esta implementaci√≥n, se probar√°n tres modelos de BERT preentrenados para evaluar su rendimiento en un conjunto de datos espec√≠fico.

#### Modelos Preentrenados Utilizados

1. **dccuchile/bert-base-spanish-wwm-cased**
   - **Descripci√≥n**: Modelo BERT preentrenado en espa√±ol con **Whole Word Masking (WWM)**, que enmascara palabras completas en lugar de subpalabras. La versi√≥n "cased" distingue entre may√∫sculas y min√∫sculas.
   - **Uso**: Adecuado para tareas en las que la distinci√≥n de may√∫sculas y min√∫sculas es relevante.

2. **bertin-project/bertin-roberta-base-spanish**
   - **Descripci√≥n**: Variante de **RoBERTa** optimizada para el espa√±ol, que elimina la tarea de predicci√≥n de la siguiente frase y utiliza enmascaramiento avanzado para mejor rendimiento en comprensi√≥n del lenguaje.
   - **Uso**: Ideal para tareas de NLP que requieren mayor capacidad de generalizaci√≥n y comprensi√≥n en espa√±ol.

3. **mrm8488/bert-base-spanish-wwm-uncased**
   - **Descripci√≥n**: Modelo BERT en espa√±ol con **Whole Word Masking** y sin distinci√≥n entre may√∫sculas y min√∫sculas ("uncased"), lo que simplifica el procesamiento en casos donde la capitalizaci√≥n no es cr√≠tica.
   - **Uso**: √ötil para an√°lisis de texto donde la capitalizaci√≥n no afecta significativamente el rendimiento.

Evaluar el rendimiento de estos modelos en tareas de clasificaci√≥n de texto para determinar cu√°l ofrece los mejores resultados en t√©rminos de precisi√≥n, recall y F1-score en el conjunto de datos seleccionado.

#### ‚≠ïÔ∏è **Implementaci√≥n** `dccuchile/bert-base-spanish-wwm-cased`

```python
from transformers import BertTokenizer, BertForSequenceClassification, TrainingArguments, Trainer
from datasets import Dataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score

# Dividir el conjunto de datos en entrenamiento y prueba
train_texts, val_texts, train_labels, val_labels = train_test_split(
    df_filtrado['Texto_Procesado'].tolist(), df_filtrado['Etiqueta_Num'].tolist(), test_size=0.2, random_state=42
)

# Cargar el tokenizer y el modelo BERT preentrenado para clasificaci√≥n multiclase
model_name = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=6)

# Funci√≥n para tokenizar los datos
def preprocess_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

# Crear un dataset de Hugging Face
train_dataset = Dataset.from_dict({"text": train_texts, "label": train_labels})
val_dataset = Dataset.from_dict({"text": val_texts, "label": val_labels})

# Aplicar la funci√≥n de preprocesamiento
train_dataset = train_dataset.map(preprocess_function, batched=True)
val_dataset = val_dataset.map(preprocess_function, batched=True)

# Definir los argumentos de entrenamiento
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
)

# Definir las m√©tricas de evaluaci√≥n
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    f1 = f1_score(labels, preds, average="weighted")
    acc = accuracy_score(labels, preds)
    return {"accuracy": acc, "f1": f1}

# Crear el Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset
)

# Entrenar el modelo
trainer.train()

# Evaluar el modelo
eval_results = trainer.evaluate()
print(f"**Resultados de la evaluaci√≥n:**\n{eval_results}")

# Funci√≥n para clasificar nuevos comentarios
def classify_comment(comment):
    inputs = tokenizer(comment, padding="max_length", truncation=True, return_tensors="pt")
    outputs = model(**inputs)
    predicted_class_id = outputs.logits.argmax(-1).item()
    categories = ["deportes", "cultura", "economia", "justicia", "politica", "salud"]
    return categories[predicted_class_id]

# Ejemplo de uso
comment_to_classify = "El hotel estuvo genial"
predicted_class = classify_comment(comment_to_classify)
print(f"El comentario '{comment_to_classify}' es de la categor√≠a: {predicted_class}")
```
