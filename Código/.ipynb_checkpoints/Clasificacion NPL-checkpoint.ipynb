#### **⭕️ Introducción** 

El presente proyecto tiene como objetivo diseñar e implementar un sistema de clasificación automática de noticias utilizando técnicas avanzadas de Procesamiento de Lenguaje Natural (NLP). A lo largo del desarrollo, se abordarán distintas etapas clave: desde la carga y exploración del conjunto de datos, centrado en noticias categorizadas en temas especializados como deportes, cultura, economía y justicia, hasta la implementación de modelos de aprendizaje profundo basados en arquitecturas RNN y LSTM. Estas etapas incluyen el preprocesamiento del texto, la filtración de datos para enfocarse en categorías relevantes, y la evaluación comparativa de los modelos mediante métricas como precisión y F1-score. Este trabajo busca no solo optimizar la clasificación de noticias, sino también explorar el impacto de distintas arquitecturas en la eficacia de sistemas aplicables a sistemas de recomendación y análisis de medios.

#### **⭕️ Núcleo Problémico**

El problema central de este proyecto radica en la dificultad para clasificar automáticamente noticias en categorías relevantes y especializadas, como deportes, cultura, economía y justicia, en un contexto donde los datos suelen incluir información heterogénea y categorías poco definidas, como “archivo”. Esta clasificación es esencial para optimizar la organización y análisis de grandes volúmenes de información en medios digitales, sistemas de recomendación y análisis de tendencias. Sin embargo, la falta de preprocesamiento adecuado, la presencia de datos irrelevantes y la selección de modelos ineficaces pueden afectar significativamente la precisión y utilidad de los resultados. Por ello, el proyecto se enfoca en abordar estos desafíos mediante un flujo sistemático de filtrado de datos, preprocesamiento textual y comparación de modelos de aprendizaje profundo, con el fin de garantizar una clasificación precisa y aplicable a temas bien definidos.

#### **⭕️ Variables con datos faltantes, Porcentaje total de valores faltantes en la base de datos.**

```python
def missing_values_per_column(df):
    """
    Esta función analiza y muestra información sobre los valores faltantes en un DataFrame de pandas.
    - Imprime el número de valores faltantes por columna.
    - Calcula y muestra el porcentaje de valores faltantes por columna.
    - Muestra el porcentaje total de valores faltantes en el DataFrame.
    - Genera un gráfico de barras que visualiza el porcentaje de valores faltantes por columna.

    Argumentos:
    df -- DataFrame de pandas que se analiza para encontrar valores faltantes.

    Retorno:
    None
    """

    # Paso 1: Encontrar columnas con valores faltantes y calcular la cantidad de valores faltantes por columna
    faltantes_por_columna = df.isnull().sum()
    
    # Paso 2: Filtrar las columnas que tienen valores faltantes y convertir los resultados a un diccionario
    data1 = faltantes_por_columna[faltantes_por_columna > 0]
    variables_con_faltantes = data1.to_dict()
    
    # Imprimir la cantidad de valores faltantes por columna en formato JSON con sangría para mejor lectura
    print('\nCant. valores faltantes por columna:',
          json.dumps(variables_con_faltantes, indent=4),
          sep="\n")
    
    # Paso 3: Calcular el porcentaje de valores faltantes por columna y formatearlo como porcentaje
    data = (data1 / len(df)) * 100
    data = data.to_dict()
    porcentaje_faltantes_por_columna = {key: f'{value:.2f}%' for key, value in data.items()}
    
    # Paso 4: Calcular el porcentaje total de valores faltantes en todo el DataFrame
    porcentaje_total_faltantes = (df.isnull().sum().sum() / df.size) * 100
    porcentaje_total_no_faltantes = (df.notnull().sum().sum() / df.size) * 100
    
    # Imprimir el porcentaje de valores faltantes por columna en formato JSON con sangría
    print('\n% valores faltantes por columna:', f'{json.dumps(porcentaje_faltantes_por_columna, indent=4)}', sep="\n")

    # Crear un gráfico de barras para visualizar el porcentaje de valores faltantes por columna
    plt.figure(figsize=(6, 3))
    plt.bar(variables_con_faltantes.keys(), variables_con_faltantes.values(), color=['blue', 'orange', 'green'])
    plt.title('Porcentaje de datos faltantes por columna')
    plt.ylabel('Porcentaje')
    plt.show()
    
    # Imprimir el porcentaje total de valores faltantes y no faltantes en el DataFrame
    print(f"\nPorcentaje total de valores faltantes en la base de datos: {porcentaje_total_faltantes:.2f}%")
    print(f"Porcentaje total de valores en la base de datos: {porcentaje_total_no_faltantes:.2f}%")

# Llamada a la función con el DataFrame 'df'
missing_values_per_column(df)
```

Además, se obtiene el primer párrafo del contenido de la noticia desde el mismo enlace para completar las columnas de info y contenido en el conjunto de datos. Este enfoque permite enriquecer el conjunto de datos con información actualizada y específica, mejorando su calidad y utilidad para análisis posteriores o modelos de machine learning.

```python
# Configuración del archivo de log
logging.basicConfig(
    filename='output_proceso_imputacion.log',  # Nombre del archivo donde se guardan los logs
    level=logging.INFO,  # Nivel de registro (INFO captura los mensajes de información estándar)
    format='%(asctime)s - %(levelname)s - %(message)s',  # Formato de los mensajes en el archivo de log
)

def get_title_and_content(url, get_element):
    """
    Esta función realiza una solicitud GET a una página web, analiza su contenido y devuelve el título o el contenido
    de la página según el parámetro 'get_element'. Los resultados y los errores se registran en un archivo de log.

    Argumentos:
    url -- URL de la página web a la que se realiza la solicitud.
    get_element -- Especifica qué parte de la página se desea obtener ('title' para el título o 'content' para el contenido).

    Retorno:
    str -- Título o contenido de la página, o 'Sin Info' si hay un error o no se encuentra la información.
    """

    # Hacer una solicitud GET a la página web
    response = requests.get(url)

    # Registrar en el archivo de log la URL de la solicitud
    logging.info("-------------------------------------------------------------------------")
    logging.info(f'[URL Noticia]: {url}')
    logging.info("-------------------------------------------------------------------------")
    
    # Verificar que la solicitud fue exitosa (código de estado 200)
    if response.status_code == 200:
        # Analizar el contenido de la página utilizando BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        response_data = ''  # Variable para almacenar el resultado de la solicitud
        
        # Si se solicita el título de la página
        if get_element == 'title':
            titulo = soup.title.string if soup.title else 'No se encontró el título'
            response_data = titulo
            # Registrar en el archivo de log el título encontrado
            logging.info(f'[Título de la página]: {titulo}')
            logging.info("-------------------------------------------------------------------------")
            
        # Si se solicita el contenido de la página
        if get_element == 'content':
            # Buscar todos los elementos <div> con la clase 'paragraph'
            parrafos_divs = soup.find_all('div', class_='paragraph')
            # Unir el texto de todos los párrafos encontrados
            contenido_parrafos = ' '.join([div.text.strip() for div in parrafos_divs if div.text.strip()]) if parrafos_divs else 'No se encontró el contenido del párrafo'
            response_data = contenido_parrafos
            # Registrar en el archivo de log una muestra del contenido encontrado (hasta 50 caracteres)
            logging.info(f'[Contenido de la noticia]: {contenido_parrafos[:50]}')
            logging.info("-------------------------------------------------------------------------")
        
        return response_data
    else:
        # Registrar en el archivo de log el error si la solicitud no fue exitosa
        logging.error(f'Error al acceder a la página. Código de estado: {response.status_code}')
        return 'Sin Info'
```

## ❇️ Preprocesamiento de Texto

Estrategías de procesamiento de texto a implementar:
- **Tokenización:** Separar el texto en palabras o tokens individuales.
- **Normalización:** Convertir el texto a minúsculas y eliminar caracteres no deseados.
- **Eliminación de Stopwords:** Quitar palabras comunes que no aportan valor semántico.

```python
# Descargar recursos de nltk si no están descargados
nltk.download('stopwords')
nltk.download('punkt')

# Definir las stopwords en inglés (puedes cambiar el idioma si es necesario)
stop_words = set(stopwords.words('english'))

# Definir la función de normalización y eliminación de stopwords
def normalizar_texto(texto):
    # Convertir a minúsculas
    texto = texto.lower()
    # Eliminar caracteres no alfabéticos
    texto = re.sub(r'[^a-z\s]', '', texto)
    # Tokenizar el texto
    palabras = word_tokenize(texto)
    # Eliminar stopwords
    palabras = [palabra for palabra in palabras if palabra not in stop_words]
    # Reconstruir el texto sin las stopwords
    return ' '.join(palabras)

# Aplicar la función de preprocesamiento al DataFrame
df_filtrado['Texto_Procesado'] = df_filtrado['contenido'].astype(str).apply(normalizar_texto)

# Tokenización y generación de secuencias
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=10000)  # Define un vocabulario máximo si es necesario
tokenizer.fit_on_texts(df_filtrado['Texto_Procesado'])
sequences = tokenizer.texts_to_sequences(df_filtrado['Texto_Procesado'])

# Calcular la longitud máxima de las secuencias
max_sequence_len = max([len(x) for x in sequences])

# Rellenar las secuencias para que tengan una longitud uniforme
sequences_padded = tf.keras.preprocessing.sequence.pad_sequences(sequences, maxlen=max_sequence_len)

# Mostrar una muestra de las secuencias procesadas
print(f"Secuencias procesadas (muestra): {sequences_padded[:5]}")
print(f"Secuencias procesadas (shape): {sequences_padded.shape}")
print(f"Dataframe (Etiquetas): {df_filtrado['Etiqueta'].unique().shape[0]}")
```

## ❇️ División del conjunto de datos

La tarea de división del conjunto de datos es fundamental para garantizar que el modelo se entrene y evalúe de manera efectiva. Para ello, se debe dividir el conjunto de datos preprocesado en dos partes: un subconjunto de entrenamiento y otro de validación. Es importante realizar esta división de forma aleatoria para evitar sesgos, pero asegurándose de que se mantenga una proporción similar de las categorías en ambos conjuntos. De esta manera, se preserva la representación de cada clase y se facilita una evaluación más precisa del rendimiento del modelo en datos no vistos, lo que contribuye a una mejor generalización.

```python
# Crear una codificación numérica para las etiquetas
# Convertimos la columna 'Etiqueta' del DataFrame 'df_filtrado' en una representación numérica.
# La función 'astype('category').cat.codes' asigna un código numérico a cada categoría única.
df_filtrado['Etiqueta_Num'] = df_filtrado['Etiqueta'].astype('category').cat.codes

# Convertir las etiquetas a formato one-hot encoding
# Calculamos el número total de clases únicas en las categorías seleccionadas.
num_classes = len(categorias_seleccionadas)
# Usamos 'tf.keras.utils.to_categorical' para convertir las etiquetas numéricas en una representación one-hot.
# Esto crea una matriz de codificación one-hot para las etiquetas de las noticias.
y = tf.keras.utils.to_categorical(df_filtrado['Etiqueta_Num'], num_classes=num_classes)

# Dividir los datos en conjuntos de entrenamiento y prueba manteniendo la proporción de categorías
# Utilizamos 'train_test_split' para dividir las características (sequences_padded) y las etiquetas (y)
# en conjuntos de entrenamiento y prueba. La opción 'test_size=0.2' indica que el 20% de los datos se usan para prueba.
# 'random_state=42' asegura la reproducibilidad de la división, y 'stratify' mantiene la proporción de clases en ambas divisiones.
X_train, X_val, y_train, y_val = train_test_split(
    sequences_padded,  # Características de entrada
    y,  # Etiquetas one-hot codificadas
    test_size=0.2,  # Proporción de datos para el conjunto de prueba
    random_state=42,  # Semilla para reproducibilidad
    stratify=df_filtrado['Etiqueta_Num']  # Mantener proporciones de categorías en la división
)

# Imprimir las formas de los conjuntos de datos
# Imprimimos las formas de los conjuntos de entrenamiento y prueba para verificar que la división se realizó correctamente.
print(f"X_train shape: {X_train.shape}")  # Forma de las características de entrenamiento
print(f"y_train shape: {y_train.shape}")  # Forma de las etiquetas de entrenamiento
print(f"X_test shape: {X_val.shape}")     # Forma de las características de prueba
print(f"y_test shape: {y_val.shape}")     # Forma de las etiquetas de prueba
```

#### **⭕️ Importante**

- **Incorporación de stratify:** Se añadió stratify=df_filtrado['Etiqueta_Num'] para que train_test_split divida los datos respetando la proporción de las categorías en Etiqueta_Num.
- **Mantener aleatoriedad:** La aleatoriedad sigue siendo controlada por random_state=42.
- **Salida balanceada:** Los conjuntos de entrenamiento y prueba mantendrán distribuciones similares para cada categoría.

## ❇️ Implementación de Modelo RNN

#### **⭕️ Procesamiento Modelo RNN**

```python
# Preparar los datos
# Extraemos las características (texto procesado) y las etiquetas de las columnas relevantes del DataFrame.
X = df_filtrado['Texto_Procesado'].values  # Características de entrada: texto procesado
y = df_filtrado['Etiqueta'].values  # Etiquetas de salida

# Tokenización y secuenciación de texto
# Se define el número máximo de palabras a considerar y la longitud máxima de las secuencias de texto.
max_words = 10000  # Número máximo de palabras a considerar en el vocabulario
max_len = 150  # Longitud máxima de la secuencia de texto

# Se crea un tokenizador que convierte el texto en secuencias de enteros, limitando el vocabulario a 'max_words'.
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(X)  # Ajustamos el tokenizador a los textos de entrenamiento

# Convertimos los textos en secuencias de números y las rellenamos para que tengan la longitud 'max_len'.
X_seq = tokenizer.texts_to_sequences(X)  # Convertir el texto en secuencias de enteros
X_pad = pad_sequences(X_seq, maxlen=max_len)  # Rellenar las secuencias a 'max_len'

# Convertir las etiquetas a categorías numéricas y hacer una codificación one-hot
# Usamos LabelEncoder para transformar las etiquetas de texto a valores numéricos.
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)  # Codificar las etiquetas numéricas
y_one_hot = to_categorical(y_encoded)  # Convertir a codificación one-hot

# Construcción del modelo RNN
embedding_dim = 100  # Dimensión de los vectores de embedding para la capa de embedding

# Se define un modelo secuencial con las siguientes capas:
model_rnn = Sequential([
    Embedding(input_dim=max_words, output_dim=embedding_dim, input_length=max_len),  # Capa de embedding
    SimpleRNN(32, return_sequences=False),  # Capa RNN con 32 unidades, sin secuencias de salida
    Dense(len(label_encoder.classes_), activation='softmax')  # Capa de salida con activación softmax para clasificación multiclase
])

# Compilación del modelo
# Se usa el optimizador 'adam' y la función de pérdida 'categorical_crossentropy' para clasificación multiclase.
model_rnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# División de datos (ejemplo, si no están divididos)
# Dividimos los datos en conjuntos de entrenamiento y validación, manteniendo la proporción de clases.
X_train, X_val, y_train, y_val = train_test_split(X_pad, y_one_hot, test_size=0.2, random_state=42, stratify=y_one_hot)

# Entrenamiento del modelo
# Entrenamos el modelo con los datos de entrenamiento, utilizando 20 épocas y un tamaño de lote de 32.
history_rnn = model_rnn.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_val, y_val))

# Evaluación del modelo
# Evaluamos el modelo en el conjunto de validación para obtener la pérdida y precisión.
loss, accuracy = model_rnn.evaluate(X_val, y_val)

# Realizar predicciones en el conjunto de validación
# Obtenemos las predicciones del modelo y convertimos las probabilidades en clases predichas.
y_pred = model_rnn.predict(X_val)
y_pred_classes = np.argmax(y_pred, axis=1)  # Obtener las clases predichas (índices de clase)
y_true = np.argmax(y_val, axis=1)  # Convertir las etiquetas one-hot a clases enteras

# Calcular recall y F1-score
# Calculamos el recall y el F1-score ponderados para evaluar la calidad de las predicciones.
recall = recall_score(y_true, y_pred_classes, average='weighted')
f1 = f1_score(y_true, y_pred_classes, average='weighted')

# Generar la matriz de confusión
# Calculamos la matriz de confusión para observar la cantidad de verdaderos positivos, falsos positivos, etc.
conf_matrix = confusion_matrix(y_true, y_pred_classes)

# Generar el reporte de clasificación
# Generamos un reporte de clasificación detallado que incluye precisión, recall y F1-score por clase.
class_report = classification_report(y_true, y_pred_classes, target_names=label_encoder.classes_)

# Imprimir resultados
# Imprimimos los resultados de la evaluación, la matriz de confusión y el reporte de clasificación.
print(f'Pérdida: {loss:.4f}, Precisión: {accuracy:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}')
print("\nMatriz de Confusión:")
print(conf_matrix)
print("\nReporte de Clasificación:")
print(class_report)
```

#### **⭕️ Mejora de Rendimiento de Modelo LSTM**

Para mejorar el rendimiento del modelo LSTM se debe evitar el sobreajuste y optimizar el entrenamiento del modelo LSTM.

#### Cambios realizados en el modelo LSTM

- **LSTM Bidireccional**: Se añadió para permitir que el modelo capture dependencias de la secuencia en ambas direcciones (izquierda a derecha y derecha a izquierda), mejorando la comprensión de la información temporal.

- **Regularización L2**: Se incorporó en las capas LSTM y densas para reducir el sobreajuste penalizando los pesos grandes, ayudando a mejorar la generalización del modelo.

- **ReduceLROnPlateau**: Se agregó como un callback para ajustar la tasa de aprendizaje automáticamente si la pérdida de validación no mejora en un número especificado de épocas, lo que puede ayudar a evitar que el modelo quede atrapado en mínimos locales.

- **Más capas y unidades**: Se incrementó la complejidad del modelo añadiendo una segunda capa LSTM y unidades de mayor capacidad para mejorar la capacidad de aprendizaje del modelo.

- **Tasa de aprendizaje personalizada**: Se estableció un valor de 0.001 en el optimizador Adam, lo cual es más común para modelos LSTM, permitiendo una mejor convergencia.

- **Callback de EarlyStopping**: Se mantuvo con una paciencia de 10 épocas para detener el entrenamiento si no hay mejora y restaurar los mejores pesos encontrados.

```python
# Configuración de parámetros para la tokenización y el modelo
max_words = 10000  # Número máximo de palabras a considerar en el vocabulario
max_len = 150      # Longitud máxima de las secuencias de entrada (máximo de palabras por muestra)

# Definición del modelo LSTM mejorado con regularización y capas bidireccionales
model_lstm = Sequential([
    # Capa de embedding para convertir palabras en vectores de características
    Embedding(input_dim=max_words, output_dim=100, input_length=max_len),
    
    # Capa LSTM bidireccional con 64 unidades, retorno de secuencias activado y regularización recurrente
    Bidirectional(LSTM(64, return_sequences=True, recurrent_dropout=0.3)),
    Dropout(0.5),  # Capa de abandono para prevenir el sobreajuste
    
    # Capa LSTM adicional con 32 unidades y sin retorno de secuencias, con regularización recurrente
    LSTM(32, return_sequences=False, recurrent_dropout=0.3),
    Dropout(0.5),  # Capa de abandono para mayor regularización
    
    # Capa densa con 128 unidades y activación ReLU
    Dense(128, activation='relu'),
    Dropout(0.5),  # Regularización adicional
    
    # Capa de salida con activación softmax para clasificación multiclase
    Dense(len(label_encoder.classes_), activation='softmax')
])

# Compilación del modelo con el optimizador 'adam', función de pérdida 'categorical_crossentropy' y métrica de 'accuracy'
model_lstm.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Configuración de callbacks para controlar el entrenamiento
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True)

# Entrenamiento del modelo con callbacks para mejorar la convergencia y guardar el mejor modelo
history_lstm_optimized = model_lstm.fit(
    X_train, y_train,
    epochs=20,  # Aumentar el número de épocas para un entrenamiento más exhaustivo
    batch_size=32,  # Tamaño del lote de entrenamiento
    validation_data=(X_val, y_val),  # Datos de validación para monitorear el desempeño
    callbacks=[early_stopping, model_checkpoint]  # Usar callbacks para control de entrenamiento
)

# Evaluación del modelo en el conjunto de validación
val_loss, val_accuracy = model_lstm.evaluate(X_val, y_val)

# Realizar predicciones en el conjunto de validación
y_pred = model_lstm.predict(X_val)
y_pred_classes = np.argmax(y_pred, axis=1)  # Convertir las predicciones one-hot a clases enteras
y_true = np.argmax(y_val, axis=1)  # Convertir las etiquetas one-hot de validación a clases enteras

# Calcular recall y F1-score para evaluar el rendimiento del modelo
recall = recall_score(y_true, y_pred_classes, average='weighted')
f1 = f1_score(y_true, y_pred_classes, average='weighted')

# Generar la matriz de confusión para evaluar la precisión de las predicciones por clase
conf_matrix = confusion_matrix(y_true, y_pred_classes)

# Generar un reporte de clasificación con detalles por clase
class_report = classification_report(y_true, y_pred_classes, target_names=label_encoder.classes_)

# Imprimir los resultados de la evaluación
print(f'Pérdida: {val_loss:.4f}, Precisión: {val_accuracy:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}')
print("\nMatriz de Confusión:")
print(conf_matrix)
print("\nReporte de Clasificación:")
print(class_report)
```

### ⭕️ **Implementación Modelo BERT** `mrm8488/bert-base-spanish-wwm-uncased`

### ⚙️ **Metodología**

#### 🖥️ Configuración del Dispositivo
- Se configura el dispositivo para utilizar la GPU si está disponible; de lo contrario, se usa la CPU.
`device = torch.device("cuda" if torch.cuda.is_available() else "cpu")`

#### 📚 Definición del Dataset Personalizado

#### 🔧 Inicialización del Tokenizador
- Se utiliza `BertTokenizer` de `transformers` para tokenizar los textos en español.
`tokenizer = BertTokenizer.from_pretrained("bert-base-spanish-wwm-uncased")`

#### 📈 Preparación de los Datos de Entrenamiento
- Se extraen y procesan los textos y etiquetas del DataFrame.
- Los textos se limitan a 512 caracteres para cumplir con la longitud máxima de entrada de BERT.
- Se verifica que todas las etiquetas estén dentro del rango permitido.

#### 🔄 Tokenización y Preparación del Dataset
- Los textos se tokenizan y se les aplica truncamiento y relleno para que tengan un tamaño uniforme de 512 tokens.
- Se crea una instancia de `NewsDataset` y un `DataLoader` para cargar los datos en lotes.

#### 🤖 Carga y Configuración del Modelo BERT
- Se carga el modelo `BertForSequenceClassification` con el número de etiquetas (`num_labels`) especificado.
`model = BertForSequenceClassification.from_pretrained("bert-base-spanish-wwm-uncased", num_labels=num_classes).to(device)`

#### ⚙️ Definición del Optimizador
- Se utiliza `AdamW` como optimizador con una tasa de aprendizaje de `5e-5`.
`optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)`

#### 🚀 Entrenamiento del Modelo
- Se entrena el modelo durante 3 épocas, calculando la pérdida y actualizando los parámetros.
- La pérdida de cada época se almacena y se imprime al final de la época.

#### 📊 Evaluación del Modelo
- Se evalúa el modelo en el conjunto de entrenamiento para obtener las métricas de rendimiento (precisión, recall, F1-score).
- Se imprime la matriz de confusión y un reporte de clasificación.

#### 📉 Visualización de la Pérdida
- Se crea una función `visualization_model` para graficar la pérdida de entrenamiento a lo largo de las épocas.
- Se llama a la función para mostrar la visualización.

### 🕹 **Codigo**

```python
# Configuración del dispositivo
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Dataset personalizado
class NewsDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}, torch.tensor(int(self.labels[idx]))

# Filtrar el DataFrame para eliminar las filas con las etiquetas "política" y "salud"
df_filtrado = df_filtrado[~df_filtrado['Etiqueta'].isin(['politica', 'salud'])]

# Actualizar las etiquetas numéricas si es necesario
df_filtrado['Etiqueta_Num'] = df_filtrado['Etiqueta'].astype('category').cat.codes

# Preparación de datos: extraer y procesar los textos y las etiquetas
train_texts = df_filtrado['Texto_Procesado'].tolist()
train_texts = [text[:512] for text in train_texts]  # Limitar cada texto a un máximo de 512 caracteres
train_labels = df_filtrado['Etiqueta_Num'].tolist()

# Validación de las etiquetas para asegurar que estén en el rango esperado
unique_labels = set(train_labels)
num_classes = len(unique_labels)
if not all(label < num_classes for label in unique_labels):
    raise ValueError(f"Encontradas etiquetas fuera del rango 0-{num_classes-1}")

# Inicialización del tokenizador BERT para tokenizar los textos
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# Tokenización de los textos con truncamiento y padding, y un máximo de 512 caracteres por texto
train_encodings = tokenizer(train_texts, truncation=True, padding=True, max_length=512)

# Preparación de los datasets y dataloaders para entrenamiento y validación
train_dataset = NewsDataset(train_encodings, train_labels)
train_size = int(0.8 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(train_dataset, [train_size, val_size])

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False)

# Carga y configuración del modelo BERT para clasificación de secuencias
model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=num_classes).to(device)

# Definición del optimizador con una tasa de aprendizaje ajustada
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)

# Entrenamiento del modelo
model.train()
train_loss_history = []  # Lista para almacenar la pérdida de cada época
val_loss_history = []  # Lista para almacenar la pérdida de validación por época
train_accuracy_history = []  # Lista para almacenar la precisión de entrenamiento por época
val_accuracy_history = []  # Lista para almacenar la precisión de validación por época

for epoch in range(3):  # Número de épocas de entrenamiento
    total_loss = 0
    correct_predictions = 0
    total_samples = 0

    # Entrenamiento
    for batch in train_loader:
        inputs, labels = batch
        inputs = {key: val.to(device) for key, val in inputs.items()}
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(**inputs, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        # Cálculo de la precisión de entrenamiento
        preds = torch.argmax(outputs.logits, axis=-1)
        correct_predictions += (preds == labels).sum().item()
        total_samples += labels.size(0)

    # Almacenar la pérdida total y la precisión de la época de entrenamiento
    train_loss_history.append(total_loss)
    train_accuracy_history.append(correct_predictions / total_samples)

    # Evaluación de la pérdida de validación y la precisión de validación
    model.eval()
    val_loss = 0
    correct_val_predictions = 0
    val_samples = 0
    with torch.no_grad():
        for batch in val_loader:
            inputs, labels = batch
            inputs = {key: val.to(device) for key, val in inputs.items()}
            labels = labels.to(device)

            outputs = model(**inputs, labels=labels)
            val_loss += outputs.loss.item()

            # Cálculo de la precisión de validación
            preds = torch.argmax(outputs.logits, axis=-1)
            correct_val_predictions += (preds == labels).sum().item()
            val_samples += labels.size(0)

    val_loss_history.append(val_loss)
    val_accuracy = correct_val_predictions / val_samples
    val_accuracy_history.append(val_accuracy)
    model.train()

    print(f"Época {epoch+1}, Pérdida de entrenamiento: {total_loss:.4f}, Precisión de entrenamiento: {train_accuracy_history[-1]:.4f}, Pérdida de validación: {val_loss:.4f}, Precisión de validación: {val_accuracy:.4f}")

# Evaluación del modelo en el conjunto de entrenamiento (o validación)
model.eval()
predictions, true_labels = [], []
with torch.no_grad():
    for batch in train_loader:
        inputs, labels = batch
        inputs = {key: val.to(device) for key, val in inputs.items()}
        labels = labels.to(device)
        outputs = model(**inputs)
        preds = torch.argmax(outputs.logits, axis=-1).cpu().numpy()
        predictions.extend(preds)
        true_labels.extend(labels.cpu().numpy())

# Cálculo de las métricas de rendimiento
accuracy = accuracy_score(true_labels, predictions)
f1 = f1_score(true_labels, predictions, average="weighted")
recall = recall_score(true_labels, predictions, average='weighted')
validation_loss = sum(val_loss_history) / len(val_loss_history)

# Impresión de resultados finales
print(f"Pérdida de validación: {validation_loss:.4f}, Precisión: {accuracy:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}")
print("\nMatriz de Confusión:")
print(confusion_matrix(true_labels, predictions))
print("\nReporte de Clasificación:")
print(classification_report(true_labels, predictions))
```

### ⭕️ **Implementación Modelo BERT** `mrm8488/bert-base-spanish-wwm-uncased` (Optimizado)

Para la optimización del modelo se practican:

1. **Regularización con Dropout**  
   Se agrega dropout en las capas de BERT para prevenir el sobreajuste.

2. **Weight Decay**  
   Se aplica regularización L2 en el optimizador `AdamW` para reducir el sobreajuste.

3. **Programación de Tasa de Aprendizaje**  
   Se utiliza un programador (`lr_scheduler`) para ajustar dinámicamente la tasa de aprendizaje durante el entrenamiento.

4. **Aumento del Tamaño del Lote**  
   Un lote más grande puede estabilizar las actualizaciones del gradiente.

5. **Aumento de Épocas**  
   Más épocas pueden ayudar a que el modelo converja mejor, aunque se debe monitorizar el sobreajuste.

6. **Gradiente Clipping**  
   Se limita el gradiente para evitar explosiones durante el entrenamiento.

7. **Validación Temprana**  
   Se agrega validación temprana para detener el entrenamiento si no hay mejora.

```python
# Configuración de hiperparámetros
num_epochs = 5  # Incrementar el número de épocas
batch_size = 16  # Aumentar el tamaño del lote
learning_rate = 2e-5  # Reducir la tasa de aprendizaje
weight_decay = 0.01  # Regularización L2

# Carga y configuración del modelo BERT para clasificación de secuencias
model_optimized = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=num_classes).to(device)

# Modificación del DataLoader con un tamaño de lote más grande
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

# Definir el optimizador con regularización L2 (Weight Decay)
optimizer = torch.optim.AdamW(model_optimized.parameters(), lr=learning_rate, weight_decay=weight_decay)

# Programación de tasa de aprendizaje
num_training_steps = len(train_loader) * num_epochs
lr_scheduler = get_scheduler("linear", optimizer=optimizer, num_warmup_steps=0, num_training_steps=num_training_steps)

# Gradiente Clipping
max_grad_norm = 1.0

# Inicializar listas para almacenar los resultados por época
train_loss_history = []
train_accuracy_history = []
val_loss_history = []
val_accuracy_history = []

# Inicializar listas para almacenar las predicciones y etiquetas para métricas finales
true_labels = []
predictions = []

# Entrenamiento con regularización
best_val_loss = float("inf")
patience = 4  # Número de épocas sin mejora para detener
early_stop_counter = 0

model_optimized.train()
for epoch in range(num_epochs):
    total_loss = 0
    correct_predictions = 0
    total_samples = 0

    # Entrenamiento
    for batch in train_loader:
        inputs, labels = batch
        inputs = {key: val.to(device) for key, val in inputs.items()}
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model_optimized(**inputs, labels=labels)
        loss = outputs.loss
        loss.backward()

        # Aplicar Gradiente Clipping
        torch.nn.utils.clip_grad_norm_(model_optimized.parameters(), max_grad_norm)

        optimizer.step()
        lr_scheduler.step()

        total_loss += loss.item()

        # Cálculo de la precisión de entrenamiento
        preds = torch.argmax(outputs.logits, axis=-1)
        correct_predictions += (preds == labels).sum().item()
        total_samples += labels.size(0)

        # Almacenar las etiquetas y predicciones para métricas finales
        true_labels.extend(labels.cpu().numpy())
        predictions.extend(preds.cpu().numpy())

    # Calcular pérdida y precisión de entrenamiento
    train_loss = total_loss / len(train_loader)
    train_accuracy = correct_predictions / total_samples

    # Almacenar resultados de entrenamiento
    train_loss_history.append(train_loss)
    train_accuracy_history.append(train_accuracy)

    # Evaluación en validación
    model_optimized.eval()
    val_loss = 0
    correct_val_predictions = 0
    val_samples = 0
    with torch.no_grad():
        for batch in val_loader:
            inputs, labels = batch
            inputs = {key: val.to(device) for key, val in inputs.items()}
            labels = labels.to(device)

            outputs = model_optimized(**inputs, labels=labels)
            val_loss += outputs.loss.item()

            preds = torch.argmax(outputs.logits, axis=-1)
            correct_val_predictions += (preds == labels).sum().item()
            val_samples += labels.size(0)

    val_loss /= len(val_loader)
    val_accuracy = correct_val_predictions / val_samples

    # Almacenar resultados de validación
    val_loss_history.append(val_loss)
    val_accuracy_history.append(val_accuracy)

    # Imprimir resultados de la época
    print(f"Época {epoch+1}, Pérdida de entrenamiento: {train_loss:.4f}, Precisión de entrenamiento: {train_accuracy:.4f}, Pérdida de validación: {val_loss:.4f}, Precisión de validación: {val_accuracy:.4f}")

    # Early Stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        early_stop_counter = 0
        torch.save(model_optimized.state_dict(), "best_model.pt")
    else:
        early_stop_counter += 1
        if early_stop_counter >= patience:
            print("Entrenamiento detenido por early stopping.")
            break

    model_optimized.train()

# Cargar el mejor modelo
model_optimized.load_state_dict(torch.load("best_model.pt"))

# Cálculo de las métricas de rendimiento
accuracy = accuracy_score(true_labels, predictions)
f1 = f1_score(true_labels, predictions, average="weighted")
recall = recall_score(true_labels, predictions, average='weighted')
validation_loss = sum(val_loss_history) / len(val_loss_history)

# Impresión de resultados finales
print(f"Pérdida de validación: {validation_loss:.4f}, Precisión: {accuracy:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}")
print("\nMatriz de Confusión:")
print(confusion_matrix(true_labels, predictions))
print("\nReporte de Clasificación:")
print(classification_report(true_labels, predictions))
```
