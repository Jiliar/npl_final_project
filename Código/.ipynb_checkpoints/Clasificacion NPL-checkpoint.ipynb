#### **⭕️ Introducción** 

El presente proyecto tiene como objetivo diseñar e implementar un sistema de clasificación automática de noticias utilizando técnicas avanzadas de Procesamiento de Lenguaje Natural (NLP). A lo largo del desarrollo, se abordarán distintas etapas clave: desde la carga y exploración del conjunto de datos, centrado en noticias categorizadas en temas especializados como deportes, cultura, economía y justicia, hasta la implementación de modelos de aprendizaje profundo basados en arquitecturas RNN y LSTM. Estas etapas incluyen el preprocesamiento del texto, la filtración de datos para enfocarse en categorías relevantes, y la evaluación comparativa de los modelos mediante métricas como precisión y F1-score. Este trabajo busca no solo optimizar la clasificación de noticias, sino también explorar el impacto de distintas arquitecturas en la eficacia de sistemas aplicables a sistemas de recomendación y análisis de medios.

#### **⭕️ Núcleo Problémico**

El problema central de este proyecto radica en la dificultad para clasificar automáticamente noticias en categorías relevantes y especializadas, como deportes, cultura, economía y justicia, en un contexto donde los datos suelen incluir información heterogénea y categorías poco definidas, como “archivo”. Esta clasificación es esencial para optimizar la organización y análisis de grandes volúmenes de información en medios digitales, sistemas de recomendación y análisis de tendencias. Sin embargo, la falta de preprocesamiento adecuado, la presencia de datos irrelevantes y la selección de modelos ineficaces pueden afectar significativamente la precisión y utilidad de los resultados. Por ello, el proyecto se enfoca en abordar estos desafíos mediante un flujo sistemático de filtrado de datos, preprocesamiento textual y comparación de modelos de aprendizaje profundo, con el fin de garantizar una clasificación precisa y aplicable a temas bien definidos.

#### **⭕️ Variables con datos faltantes, Porcentaje total de valores faltantes en la base de datos.**

```python
def missing_values_per_column(df):
    # Escribir las variables que tienen datos faltantes, además, proporcionar el porcentaje total de valores faltantes en la base de datos.
    # Paso 1: Encontrar columnas con valores faltantes
    faltantes_por_columna = df.isnull().sum()
    
    # Paso 2: Filtrar columnas que tienen valores faltantes
    data1 = faltantes_por_columna[faltantes_por_columna > 0]
    variables_con_faltantes = data1.to_dict()
    
    print('\nCant. valores faltantes por columna:',
          json.dumps(variables_con_faltantes, indent=4),
          sep="\n")
    
    # Paso 3: Calcular el porcentaje de valores faltantes por columna
    data = (data1 / len(df)) * 100
    data = data.to_dict()
    porcentaje_faltantes_por_columna = {key: f'{value:.2f}%' for key, value in data.items()}
    
    # Paso 4: Calcular el porcentaje total de valores faltantes en el DataFrame
    porcentaje_total_faltantes = (df.isnull().sum().sum() / df.size) * 100
    porcentaje_total_no_faltantes = (df.notnull().sum().sum() / df.size) * 100
    
    # Resultados
    print('\n% valores faltantes por columna:', f'{json.dumps(porcentaje_faltantes_por_columna, indent=4)}', sep="\n")

    # Crear gráfico de barras
    plt.figure(figsize=(6, 3))
    plt.bar(variables_con_faltantes.keys(), variables_con_faltantes.values(), color=['blue', 'orange', 'green'])
    plt.title('Porcentaje de datos faltantes por columna')
    plt.ylabel('Porcentaje')
    plt.show()
    
    print(f"\nPorcentaje total de valores faltantes en la base de datos: {porcentaje_total_faltantes:.2f}%")
    print(f"Porcentaje total de valores en la base de datos: {porcentaje_total_no_faltantes:.2f}%")

missing_values_per_column(df)
```

#### **⭕️ Imputación por extracción de datos**

El proceso de imputación de datos en este caso implica acceder a las URLs de las noticias y extraer información relevante para llenar los datos faltantes. Para aquellas noticias que no tienen título, se accede al enlace correspondiente y se extrae el título de la página. Además, se obtiene el primer párrafo del contenido de la noticia desde el mismo enlace para completar las columnas de info y contenido en el conjunto de datos. Este enfoque permite enriquecer el conjunto de datos con información actualizada y específica, mejorando su calidad y utilidad para análisis posteriores o modelos de machine learning.

```python
# Configurar el archivo de log
logging.basicConfig(
    filename='output_proceso_imputacion.log',  # Nombre del archivo de log
    level=logging.INFO,  # Nivel de registro (INFO captura las impresiones estándar)
    format='%(asctime)s - %(levelname)s - %(message)s',  # Formato del mensaje
)

def get_title_and_content(url, get_element):
    # Hacer una solicitud GET a la página web
    response = requests.get(url)

    logging.info("-------------------------------------------------------------------------")
    logging.info(f'[URL Noticia]: {url}')
    logging.info("-------------------------------------------------------------------------")
    
    # Verificar que la solicitud fue exitosa
    if response.status_code == 200:
        # Analizar el contenido de la página con BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        response_data = ''
        
        if get_element == 'title':
            # Buscar el título de la página
            titulo = soup.title.string if soup.title else 'No se encontró el título'
            response_data = titulo
            logging.info(f'[Título de la página]: {titulo}')
            logging.info("-------------------------------------------------------------------------")
            
        if get_element == 'content':
            # Buscar todos los <div> con la clase 'paragraph'
            parrafos_divs = soup.find_all('div', class_='paragraph')
            contenido_parrafos = ' '.join([div.text.strip() for div in parrafos_divs if div.text.strip()]) if parrafos_divs else 'No se encontró el contenido del párrafo'
            response_data = contenido_parrafos
            logging.info(f'[Contenido de la noticia]: {contenido_parrafos[:50]}')
            logging.info("-------------------------------------------------------------------------")
        
        return response_data
    else:
        logging.error(f'Error al acceder a la página. Código de estado: {response.status_code}')
        return 'Sin Info'
```

#### **⭕️ Imputación de Titulos de noticias**

```python
flag_imputer = False # Para imputar valores coloque esta badenra en True

if flag_imputer:
    df_new = df.copy()
    for index, row in df_new.iterrows():
        if pd.isnull(row['Título']) or row['Título'] == '':
            df_new.at[index, 'Título'] = get_title_and_content(df_new.at[index, 'Enlaces'], 'title')
    
    missing_values_per_column(df_new)
```

## ❇️ Preprocesamiento de Texto

Estrategías de procesamiento de texto a implementar:
- **Tokenización:** Separar el texto en palabras o tokens individuales.
- **Normalización:** Convertir el texto a minúsculas y eliminar caracteres no deseados.
- **Eliminación de Stopwords:** Quitar palabras comunes que no aportan valor semántico.

```python
# Descargar recursos de nltk si no están descargados
nltk.download('stopwords')
nltk.download('punkt')

# Definir las stopwords en inglés (puedes cambiar el idioma si es necesario)
stop_words = set(stopwords.words('english'))

# Definir la función de normalización y eliminación de stopwords
def normalizar_texto(texto):
    # Convertir a minúsculas
    texto = texto.lower()
    # Eliminar caracteres no alfabéticos
    texto = re.sub(r'[^a-z\s]', '', texto)
    # Tokenizar el texto
    palabras = word_tokenize(texto)
    # Eliminar stopwords
    palabras = [palabra for palabra in palabras if palabra not in stop_words]
    # Reconstruir el texto sin las stopwords
    return ' '.join(palabras)

# Aplicar la función de preprocesamiento al DataFrame
df_filtrado['Texto_Procesado'] = df_filtrado['contenido'].astype(str).apply(normalizar_texto)

# Tokenización y generación de secuencias
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=10000)  # Define un vocabulario máximo si es necesario
tokenizer.fit_on_texts(df_filtrado['Texto_Procesado'])
sequences = tokenizer.texts_to_sequences(df_filtrado['Texto_Procesado'])

# Calcular la longitud máxima de las secuencias
max_sequence_len = max([len(x) for x in sequences])

# Rellenar las secuencias para que tengan una longitud uniforme
sequences_padded = tf.keras.preprocessing.sequence.pad_sequences(sequences, maxlen=max_sequence_len)

# Mostrar una muestra de las secuencias procesadas
print(f"Secuencias procesadas (muestra): {sequences_padded[:5]}")
print(f"Secuencias procesadas (shape): {sequences_padded.shape}")
print(f"Dataframe (Etiquetas): {df_filtrado['Etiqueta'].unique().shape[0]}")
```

## ❇️ División del conjunto de datos

La tarea de división del conjunto de datos es fundamental para garantizar que el modelo se entrene y evalúe de manera efectiva. Para ello, se debe dividir el conjunto de datos preprocesado en dos partes: un subconjunto de entrenamiento y otro de validación. Es importante realizar esta división de forma aleatoria para evitar sesgos, pero asegurándose de que se mantenga una proporción similar de las categorías en ambos conjuntos. De esta manera, se preserva la representación de cada clase y se facilita una evaluación más precisa del rendimiento del modelo en datos no vistos, lo que contribuye a una mejor generalización.

```python
# Crear una codificación numérica para las etiquetas
df_filtrado['Etiqueta_Num'] = df_filtrado['Etiqueta'].astype('category').cat.codes

# Convertir las etiquetas a formato one-hot encoding
num_classes = len(categorias_seleccionadas)
y = tf.keras.utils.to_categorical(df_filtrado['Etiqueta_Num'], num_classes=num_classes)

# Dividir los datos en conjuntos de entrenamiento y prueba manteniendo la proporción de categorías
X_train, X_val, y_train, y_val = train_test_split(
    sequences_padded,
    y,
    test_size=0.2,
    random_state=42,
    stratify=df_filtrado['Etiqueta_Num']  # Asegurar proporción similar en ambas divisiones
)

# Imprimir las formas de los conjuntos
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_val.shape}")
print(f"y_test shape: {y_val.shape}")
```

#### **⭕️ Importante**

- **Incorporación de stratify:** Se añadió stratify=df_filtrado['Etiqueta_Num'] para que train_test_split divida los datos respetando la proporción de las categorías en Etiqueta_Num.
- **Mantener aleatoriedad:** La aleatoriedad sigue siendo controlada por random_state=42.
- **Salida balanceada:** Los conjuntos de entrenamiento y prueba mantendrán distribuciones similares para cada categoría.

#### **⭕️ Procesamiento Modelo RNN Seleccionado**

```python
# Preparar los datos
X = df_filtrado['Texto_Procesado'].values
y = df_filtrado['Etiqueta'].values

# Tokenización y secuenciación de texto
max_words = 5000  # Número máximo de palabras a considerar
max_len = 150  # Longitud máxima de la secuencia de texto

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(X)
X_seq = tokenizer.texts_to_sequences(X)
X_pad = pad_sequences(X_seq, maxlen=max_len)

# Convertir las etiquetas a categorías numéricas y hacer una codificación one-hot
#from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_one_hot = to_categorical(y_encoded)

# Construcción del modelo RNN
embedding_dim = 100  # Dimensión de los vectores de embedding

model_rnn = Sequential([
    Embedding(input_dim=max_words, output_dim=embedding_dim, input_length=max_len),
    SimpleRNN(32, return_sequences=False),
    Dense(len(label_encoder.classes_), activation='softmax')  # Capa de salida con activación softmax para clasificación multiclase
])

# Compilación del modelo
model_rnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# División de datos (ejemplo, si no están divididos)
X_train, X_val, y_train, y_val = train_test_split(X_pad, y_one_hot, test_size=0.2, random_state=42, stratify=y_one_hot)

# Entrenamiento del modelo
history_rnn = model_rnn.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_val, y_val))

# Evaluación del modelo
loss, accuracy = model_rnn.evaluate(X_val, y_val)

# Realizar predicciones en el conjunto de validación
y_pred = model_rnn.predict(X_val)
y_pred_classes = np.argmax(y_pred, axis=1)  # Obtener las clases predichas
y_true = np.argmax(y_val, axis=1)  # Convertir las etiquetas one-hot a clases enteras

# Calcular recall y F1-score
recall = recall_score(y_true, y_pred_classes, average='weighted')
f1 = f1_score(y_true, y_pred_classes, average='weighted')

# Generar la matriz de confusión
conf_matrix = confusion_matrix(y_true, y_pred_classes)

# Generar el reporte de clasificación
class_report = classification_report(y_true, y_pred_classes, target_names=label_encoder.classes_)

# Imprimir resultados
print(f'Pérdida: {loss:.4f}, Precisión: {accuracy:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}')
print("\nMatriz de Confusión:")
print(conf_matrix)
print("\nReporte de Clasificación:")
print(class_report)
```

#### **⭕️ Visualización de Procesamiento Modelo RNN**

```python
def visualization_model(history):
    # Graficar la precisión y la pérdida del modelo durante el entrenamiento
    plt.figure(figsize=(12, 5))
    
    # Gráfica de la pérdida
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Pérdida de entrenamiento')
    plt.plot(history.history['val_loss'], label='Pérdida de validación')
    plt.title('Pérdida durante el entrenamiento')
    plt.xlabel('Épocas')
    plt.ylabel('Pérdida')
    plt.legend()
    
    # Gráfica de la precisión
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Precisión de entrenamiento')
    plt.plot(history.history['val_accuracy'], label='Precisión de validación')
    plt.title('Precisión durante el entrenamiento')
    plt.xlabel('Épocas')
    plt.ylabel('Precisión')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    # Crear un DataFrame con la información de épocas y pérdidas
    df_loss = pd.DataFrame({
        'Épocas': range(1, len(history.history['loss']) + 1),
        'Pérdida de entrenamiento': history.history['loss'],
        'Pérdida de validación': history.history['val_loss']
    })
    
    # Mostrar la tabla
    print(df_loss)

visualization_model(history_rnn)
```

## ❇️ Implementación de Modelo LSTM

En la fase de implementación del modelo LSTM, se procederá a definir y construir una red neuronal con capas de embedding y LSTM, diseñada para capturar dependencias a largo plazo en los datos textuales. Esta etapa incluye la configuración de los hiperparámetros y el entrenamiento del modelo utilizando el conjunto de entrenamiento, aplicando técnicas de regularización si es necesario para evitar el sobreajuste. Una vez entrenado, se evaluará el rendimiento del modelo con el conjunto de validación y se comparará con el modelo RNN previamente desarrollado, con el fin de seleccionar el enfoque más eficiente y eficaz para la clasificación de noticias en las categorías definidas.

Para mejorar el rendimiento del modelo LSTM se debe evitar el sobreajuste y optimizar el entrenamiento del modelo LSTM.

#### Cambios realizados en el modelo LSTM

- **LSTM Bidireccional**: Se añadió para permitir que el modelo capture dependencias de la secuencia en ambas direcciones (izquierda a derecha y derecha a izquierda), mejorando la comprensión de la información temporal.

- **Regularización L2**: Se incorporó en las capas LSTM y densas para reducir el sobreajuste penalizando los pesos grandes, ayudando a mejorar la generalización del modelo.

- **ReduceLROnPlateau**: Se agregó como un callback para ajustar la tasa de aprendizaje automáticamente si la pérdida de validación no mejora en un número especificado de épocas, lo que puede ayudar a evitar que el modelo quede atrapado en mínimos locales.

- **Más capas y unidades**: Se incrementó la complejidad del modelo añadiendo una segunda capa LSTM y unidades de mayor capacidad para mejorar la capacidad de aprendizaje del modelo.

- **Tasa de aprendizaje personalizada**: Se estableció un valor de 0.001 en el optimizador Adam, lo cual es más común para modelos LSTM, permitiendo una mejor convergencia.

- **Callback de EarlyStopping**: Se mantuvo con una paciencia de 10 épocas para detener el entrenamiento si no hay mejora y restaurar los mejores pesos encontrados.

```python
# Configuración de parámetros
max_words = 10000  # Número máximo de palabras en el vocabulario
max_len = 150      # Longitud máxima de las secuencias de entrada

# Definición del modelo LSTM mejorado
model_lstm = Sequential([
    Embedding(input_dim=max_words, output_dim=100, input_length=max_len),
    Bidirectional(LSTM(64, return_sequences=True, recurrent_dropout=0.3)),  # LSTM bidireccional con regularización recurrente
    Dropout(0.5),  # Regularización para prevenir el sobreajuste
    LSTM(32, return_sequences=False, recurrent_dropout=0.3),  # Capa LSTM adicional
    Dropout(0.5),  # Otra capa de regularización
    Dense(128, activation='relu'),
    Dropout(0.5),  # Regularización adicional
    Dense(len(label_encoder.classes_), activation='softmax')  # Capa de salida para clasificación multiclase
])

# Compilación del modelo con un optimizador ajustado
model_lstm.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Configuración de callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True)

# Entrenamiento del modelo
history_lstm_optimized = model_lstm.fit(
    X_train, y_train,
    epochs=20,  # Aumentar el número de épocas para un mejor entrenamiento
    batch_size=32,
    validation_data=(X_val, y_val),
    callbacks=[early_stopping, model_checkpoint]  # Usar callbacks
)

# Evaluación del modelo con el conjunto de validación
val_loss, val_accuracy = model_lstm.evaluate(X_val, y_val)

# Realizar predicciones en el conjunto de validación
y_pred = model_lstm.predict(X_val)
y_pred_classes = np.argmax(y_pred, axis=1)  # Obtener las clases predichas
y_true = np.argmax(y_val, axis=1)  # Convertir las etiquetas one-hot a clases enteras

# Calcular recall y F1-score
recall = recall_score(y_true, y_pred_classes, average='weighted')
f1 = f1_score(y_true, y_pred_classes, average='weighted')

# Generar la matriz de confusión
conf_matrix = confusion_matrix(y_true, y_pred_classes)

# Generar el reporte de clasificación
class_report = classification_report(y_true, y_pred_classes, target_names=label_encoder.classes_)

# Imprimir resultados
print(f'Pérdida: {val_loss:.4f}, Precisión: {val_accuracy:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}')
print("\nMatriz de Confusión:")
print(conf_matrix)
print("\nReporte de Clasificación:")
print(class_report)
```

Para abordar tareas de clasificación de texto en español, se utilizan modelos de lenguaje avanzados como BERT, que permiten capturar el contexto de las palabras y mejorar la comprensión del texto. 1. **dccuchile/bert-base-spanish-wwm-cased**

3. **mrm8488/bert-base-spanish-wwm-uncased**
   - **Descripción**: Modelo BERT en español con **Whole Word Masking** y sin distinción entre mayúsculas y minúsculas ("uncased"), lo que simplifica el procesamiento en casos donde la capitalización no es crítica.
   - **Uso**: Útil para análisis de texto donde la capitalización no afecta significativamente el rendimiento.

Evaluar el rendimiento de estos modelos en tareas de clasificación de texto para determinar cuál ofrece los mejores resultados en términos de precisión, recall y F1-score en el conjunto de datos seleccionado.

#### ⭕️ **Implementación** `dccuchile/bert-base-spanish-wwm-cased`

```python
from transformers import BertTokenizer, BertForSequenceClassification, TrainingArguments, Trainer
from datasets import Dataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score

# Dividir el conjunto de datos en entrenamiento y prueba
train_texts, val_texts, train_labels, val_labels = train_test_split(
    df_filtrado['Texto_Procesado'].tolist(), df_filtrado['Etiqueta_Num'].tolist(), test_size=0.2, random_state=42
)

# Cargar el tokenizer y el modelo BERT preentrenado para clasificación multiclase
model_name = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=6)

# Función para tokenizar los datos
def preprocess_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

# Crear un dataset de Hugging Face
train_dataset = Dataset.from_dict({"text": train_texts, "label": train_labels})
val_dataset = Dataset.from_dict({"text": val_texts, "label": val_labels})

# Aplicar la función de preprocesamiento
train_dataset = train_dataset.map(preprocess_function, batched=True)
val_dataset = val_dataset.map(preprocess_function, batched=True)

# Definir los argumentos de entrenamiento
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
)

# Definir las métricas de evaluación
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    f1 = f1_score(labels, preds, average="weighted")
    acc = accuracy_score(labels, preds)
    return {"accuracy": acc, "f1": f1}

# Crear el Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset
)

# Entrenar el modelo
trainer.train()

# Evaluar el modelo
eval_results = trainer.evaluate()
print(f"**Resultados de la evaluación:**\n{eval_results}")

# Función para clasificar nuevos comentarios
def classify_comment(comment):
    inputs = tokenizer(comment, padding="max_length", truncation=True, return_tensors="pt")
    outputs = model(**inputs)
    predicted_class_id = outputs.logits.argmax(-1).item()
    categories = ["deportes", "cultura", "economia", "justicia", "politica", "salud"]
    return categories[predicted_class_id]

# Ejemplo de uso
comment_to_classify = "El hotel estuvo genial"
predicted_class = classify_comment(comment_to_classify)
print(f"El comentario '{comment_to_classify}' es de la categoría: {predicted_class}")
```
